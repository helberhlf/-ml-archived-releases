### 1.2.1. O que é churn de clientes?

Segundo <a href="https://blog.hubspot.com/service/what-is-customer-churn">(AMARESAN, 2021)</a> refina que a rotatividade de clientes é uma das métricas mais importantes para medir um negócio (para uma empresa) em crescimento. Embora não seja a métrica mais feliz, é um número que pode dar à sua empresa a dura verdade sobre a retenção de clientes.

É difícil medir o sucesso se você não medir os inevitáveis fracassos também, isso ressalva a seguinte frase do autor **Winston Churchill** **<font color="red">“O sucesso não é definitivo, o fracasso não é fatal: é a coragem de continuar que conta.”</font>**.Mas, simplesmente não é realista quando você está tentando fazer com que 100% de seus clientes permaneçam com sua empresa. É daí que vem o **churn do cliente.**

<font color="red">Nesse mesmo sentido, em poucas palavras  rotatividade de clientes, **também conhecida como perda de clientes**, é quando um cliente opta por parar de usar seus produtos ou serviços. Com efeito, é quando um **cliente deixa de ser um cliente.<font color="red">**

<font color="red">O Churn de clientes é medido usando a taxa de churn de clientes. Esse é o número de pessoas que deixaram de ser clientes durante um determinado período de tempo, como um ano, um mês ou um trimestre financeiro.Em poucas palavras,**a rotatividade de clientes é a porcentagem de clientes que sua empresa perdeu em um determinado período de tempo.<font color="red">**

### 1.2.2. Como calcular o Customer Churn ?

Existem muitas maneiras de calcular o Customer Churn, uma delas é dividir o número de clientes que saem da empresa em um determinado intervalo de tempo pelo número de clientes que estavam presentes no início do período.A fórmula para calcular o Customer Churn é simples. Vejamos:

$$ Taxa​​de​​Churn = \frac {Números​​de​​clientes​​que​​cancelaram} {Números​​de​​clientes​​ativos​​no​​início​​do ​​período}​​x​​100 $$

**Veja um exemplo de taxa de rotatividade de clientes:**

A empresa X perdeu **200 clientes** <a href="https://www.investopedia.com/terms/b/btoc.asp">**Business to Consumer - B2C</a>** em um período mensal.Mas, ela tinha **4.000 clientes** no início do mês e terminou com **3.800**.Usando a fórmula temos:

$$ Taxa​​de ​​Churn= \frac {200​​Clientes​​Perdidos}{4.000​​Clientes​​no​​Início​​do​​Período​​de​​Tempo ​​Escolhido} x 100 = 5\% $$
> Ou seja, a empresa do exemplo teve uma **taxa de cancelamento de 5%** no mês analisado.</font>

### Por que a taxa de rotatividade de clientes é importante ?

Em suma, se a sua empresa não consegue manter os clientes, é quase impossível desenvolver um negócio sustentável. Mas quando a **taxa de churn** é muito alta ou mostra uma **tendência crescente ao longo do tempo**, a empresa  deve rapidamente começa a agir.

Muitas vezes, é um híbrido dos dois. ### 2.2. Termo-chave: previsão
No uso comum, **previsão significa antecipar um evento futuro**. Em Data science, de forma mais geral,previsão significa **estimar um valor desconhecido**. Esse valor poderia ser algo no futuro (no uso comum, previsão verdadeira), mas também pode ser algo no presente ou no passado. De fato, como a mineração de dados geralmente lida com dados históricos, os modelos costumam ser construídos e testados por meio do **uso de eventos do passado**. Modelos preditivos para pontuação de crédito estimam a probabilidade de um potencial cliente ser inadimplente. Modelos preditivos para filtragem de spam estimam se determinado e-mail é spam. Modelos preditivos para a detecção de fraude julgam se uma conta foi defraudada. O segredo é que o modelo seja destinado para **estimar  um valor desconhecido**.**(FAWCETT and PROVOST,  2016)**.


### 2.2. Termo-chave: Indução e Dedução

A indução pode ser comparada com a **dedução**.Porém, começa com regras gerais e fatos específicos, e cria outros fatos específicos a partir deles.A utilização dos nossos modelos pode ser considerada um procedimento de dedução (probabilística).**(FAWCETT and PROVOST, 2016)**.

A criação de modelos a partir de dados é conhecida como modelo de indução.A indução é um termo da filosofia que se refere a generalização a partir de casos específicos até regras gerais (ou leis ou verdades). Nossos modelos são regras gerais no sentido estatístico (eles normalmente não se mantêm 100% das vezes; com frequência, nem chegam perto), e o procedimento que cria o modelo a partir de dados é chamado algoritmo de indução ou aprendiz.**(FAWCETT and PROVOST, 2016)**.

- LabelEncoder aprende classes_

- OrdinalEncoder aprende categories_


Em suma, as técnicas  basicamente fazem a mesma coisa (como você pode ver), mas a diferença é que **OrdinalEncoder** é usado para transformar o recurso e **LabelEncoder** é usado para transformar a variável de destino. Com base nos conceitos citados acima, a função adiante é pouco complicada, ela basicamente retorna o nomes das variáveis dos dados transformados ao final da aplicação.O código a seguir é sobre a <a href="https://github.com/scikit-learn/scikit-learn/issues/12525">Cannot get feature names after ColumnTransformer #12525</a>

```python
from skimpy import clean_columns
from sklearn.utils.validation import check_is_fitted

def get_column_names_from_ColumnTransformer(column_transformer, clean_column_names=True, verbose=True):  

    """
    Reference: Kyle Gilde: https://github.com/kylegilde/Kaggle-Notebooks/blob/master/Extracting-and-Plotting-Scikit-Feature-Names-and-Importances/feature_importance.py
    Description: Get the column names from the a ColumnTransformer containing transformers & pipelines
    Parameters
    ----------
    verbose: Bool indicating whether to print summaries. Default set to True.
    Returns
    -------
    a list of the correct feature names
    Note:
    If the ColumnTransformer contains Pipelines and if one of the transformers in the Pipeline is adding completely new columns,
    it must come last in the pipeline. For example, OneHotEncoder, MissingIndicator & SimpleImputer(add_indicator=True) add columns
    to the dataset that didn't exist before, so there should come last in the Pipeline.
    Inspiration: https://github.com/scikit-learn/scikit-learn/issues/12525
    """

    assert isinstance(column_transformer, ColumnTransformer), "Input isn't a ColumnTransformer"
    
    check_is_fitted(column_transformer)

    new_feature_names, transformer_list = [], []

    for i, transformer_item in enumerate(column_transformer.transformers_): 
        transformer_name, transformer, orig_feature_names = transformer_item
        orig_feature_names = list(orig_feature_names)

        if len(orig_feature_names) == 0:
            continue

        if verbose: 
            print(f"\n\n{i}.Transformer/Pipeline: {transformer_name} {transformer.__class__.__name__}\n")
            print(f"\tn_orig_feature_names:{len(orig_feature_names)}")

        if transformer == 'drop':
            continue

        if isinstance(transformer, Pipeline):
            # if pipeline, get the last transformer in the Pipeline
            transformer = transformer.steps[-1][1]

        if hasattr(transformer, 'get_feature_names_out'):
            if 'input_features' in transformer.get_feature_names_out.__code__.co_varnames:
                names = list(transformer.get_feature_names_out(orig_feature_names))
            else:
                names = list(transformer.get_feature_names_out())
        elif hasattr(transformer, 'get_feature_names'):
            if 'input_features' in transformer.get_feature_names.__code__.co_varnames:
                names = list(transformer.get_feature_names(orig_feature_names))
            else:
                names = list(transformer.get_feature_names())

        elif hasattr(transformer,'indicator_') and transformer.add_indicator:
            # is this transformer one of the imputers & did it call the MissingIndicator?

            missing_indicator_indices = transformer.indicator_.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]
            names = orig_feature_names + missing_indicators

        elif hasattr(transformer,'features_'):
            # is this a MissingIndicator class? 
            missing_indicator_indices = transformer.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]

        else:

            names = orig_feature_names

        if verbose: 
            print(f"\tn_new_features:{len(names)}")
            print(f"\tnew_features: {names}\n")

        new_feature_names.extend(names)
        transformer_list.extend([transformer_name] * len(names))

    transformer_list, column_transformer_features = transformer_list, new_feature_names

    if clean_column_names:
        new_feature_names = list(clean_columns(pd.DataFrame(columns=new_feature_names)).columns)
    
    return new_feature_names
```

### 5.4.1. Assimetria  (Skewness, em inglês)

<font color ="red">É o grau de desvio ou afastamento da simetria de uma distribuição. Quando a curva é simétrica, a média, a mediana e a moda coincidem, num mesmo ponto, de ordenada máxima, havendo um perfeito equilíbrio na distribuição. Quando o equilíbrio não acontece, isto é, a média, a mediana e a moda recaem em pontos diferentes da distribuição esta será assimétrica; enviesada a direita ou esquerda.A figura a seguir simplifica  esse conceito </font>





Em uma outra definição mais abrangente **Skewness** é uma medida da assimetria da distribuição de probabilidade de uma variável aleatória de valor real sobre sua média. O valor da assimetria pode ser positivo, negativo ou indefinido.

O coeficiente de assimetria é o que nos permite dizer se uma determinada distribuição é assimétrica ou não em torno da média. Em  poucas palavras é quanto a sua curva de frequência se desvia ou se afasta da posição simétrica.

***
- **Sk = 0: dados simétricos**. Tanto a cauda do lado direito quanto a do lado esquerdo da função densidade de probabilidade são iguais.

   + **Distribuição Assimétrica:** média = mediana = moda
   
    
- **Sk < 0: Assimetria Negativa:** Distribuição assimétrica Negativa ou enviesada a esquerda - quando os valores se concentram na extremidade superior da escala e se distribuem gradativamente em direção à extremidade inferior.Ou seja, cauda esquerda é mais longa, assim sendo a cauda do lado esquerdo da função densidade de probabilidade é maior que a do lado direito.



   + **Assimetria negativa:** moda > mediana > média 
   
---
    
- **Sk > 0: Assimetria Positiva**.Distribuição assimétrica **Positiva** ou enviesada a direita quando os valores se concentram na extremidade inferior da escala e se distribuem gradativamente em direção à extremidade superior.Ou seja, cauda direita é mais longa, refinando que a cauda do lado direito da função densidade de probabilidade é maior do que a cauda do lado esquerdo.



  + **Assimetria positiva:** média > mediana > moda


### 5.4.2. Kurtosis ou Curtose


Coeficiente Assimetria de **Curtose** é uma medida que caracteriza o achatamento (é o grau de achatamento de uma distribuição, em relação a distribuição normal).Um dos coeficientes mais utilizados para medir o grau de achatamento de uma distribuição é o coeficiente percentílico de curtose, ou
simplesmente coeficiente de curtose (k), calculado a partir do intervalo interquartil dos percentis de ordem 10 e 90.

A medida de curtose nos indica a forma da curva de distribuição em relação ao seu achatamento.

A distribuição dos dados pode ser classificada em três classes:

-  **Mesocúrtica:** A distribuição  apresenta uma curva de frequências idêntica a da distribuição Normal.

- **Leptocúrtica:** A Distribuição apresenta uma curva de freqüência mais fechada que a normal (ou mais aguda em sua parte superior)
- **Platicúrtica:** A Distribuição apresenta uma curva de freqüência mais aberta que a normal (ou mais achatada na sua parte superior)


 
---

Coeficiente Percentílico de Curtose:


$$ C = \frac{Q3 - Q1}{ 2 (D9 - D1)}$$ 

**Relativamente a curva normal, temos; C = 0,263 Assim:**

-  **C = 0,263 ⇒** Curva Mesocúrtica
-  **C > 0,263 ⇒** Curva Leptocúrtica
-  **C < 0,263 ⇒** Curva Platicúrtica


**Percentil:** é o ponto da distribuição dos resultados ordenados da amostra(por ordem crescente dos dados) em 100 partes de igual amplitude.Por exemplo um resultado no percentil 90 significa que 90% dos resultados se situam nesse ponto ou abaixo dele, é maneira mais fácil  de informar a posição relativa.


Em suma, quando a Assimetria e Curtose são zero (uma situação que os pesquisadores dificilmente encontrarão), o padrão de respostas é considerado uma distribuição normal. Uma diretriz geral para assimetria é que, se o número for maior que **+1 ou menor que - 1, esta é uma indicação de uma distribuição substancialmente distorcida**.Para curtose, a diretriz geral é que, se o número for maior que **+1, a distribuição é muito alta**. Da mesma forma, uma curtose menor que -1 indica uma distribuição que é muito plana.As  Distribuições mostrada pela Assimetria e Curtose que excedem essas diretrizes são consideradas não normais.

Mediante disso, quando os dados estão distorcidos, a região da cauda pode se comportar como um valor atípico para o modelo estatístico, e os valores atípicos afetam o desempenho do modelo, especialmente os modelos baseados em regressão. Alguns modelos estatísticos são resistentes a discrepâncias, como os modelos baseados em árvore, mas limitarão a possibilidade de experimentar outros modelos. Portanto, há uma necessidade de transformar os dados distorcidos para perto o suficiente de uma distribuição Normal .

Observa-se que (PORTELLA, 2018) refina muitas aplicações precisam definir se uma determinada observação pertence à mesma distribuição das demais observações (é um inlier) ou se ela deve ser considerada distinta das demais observaçes (é um outlier) <a href ="https://scikit-learn.org/stable/modules/outlier_detection.html">(Sklearn)</a>. ### 6.3.1. O que são Percentiles

**Segundo (ANDREW e BRUCE, 2019) uma abordagem diferente para estimar dispersão é com base na observação da distribuição dos dados classificados. A estatística baseada em dados classificados (ordenados) é chamada de estatística ordinal. A medida mais básica é a amplitude: a diferença entre o maior e o menor número. É bom conhecer os valores máximos e mínimos, que são úteis na identificação dos outliers, mas a amplitude é extremamente sensível aos outliers e não é muito útil como medida geral da dispersão dos dados. Para evitar a sensibilidade aos outliers, podemos observar a amplitude dos dados depois da exclusão dos valores de cada ponta. Formalmente, esses tipos de estimativas são baseados em diferenças entre percentis.**<font color="red"> Em um conjunto de dados, o percentil P é um valor em que ao menos P por cento dos valores assume esse valor ou menos, e ao menos (100 – P) por cento dos valores assume esse valor ou mais. Por exemplo, para encontrar o 80° percentil, classifique os dados. Então, começando com o menor valor, avance 80% em direção ao maior valor. Note que a mediana é a mesma coisa que o 50° percentil. O percentil é basicamente o mesmo que um quantil, com quantis indexados por frações (então o quantil .8 é igual ao 80° percentil). Uma medida comum da variabilidade é a diferença entre o 25° percentil e o 75° percentil, chamada de amplitude interquartil (ou IQR).</font>

```python

'''A função retorna um vetor booleano: True se o elemento for um outlier. Falso, caso contrário.
Agora, para substituir os outliers superiores e inferiores, vamos escrever outra pequena função e aplicá-la em todo o dataframe:'''
def outliers(x): 
       return np.abs(x- x.median()) > 1.5*(x.quantile(0.95)-x.quantile(0.05))
    
'''Substitua os valores outliers mais altos pelo percentil 95 e os valores outliers mais baixos pelo percentil 5'''    
def replace(x):
    out = x[outliers(x)] # é vector
    return x.replace(to_replace = [out.min(), out.max()],
                    value = (np.percentile(x,5), np.percentile(x,95)))
```

**<a href="https://numpy.org/doc/stable/reference/generated/numpy.argpartition.html" target="blank">numpy.argpartition()**</a> A função é usada para criar uma cópia particionada indireta do array de entrada com seus elementos reorganizados de tal forma que o valor do elemento na k-ésima posição esteja na posição que estaria em um array ordenado. Todos os elementos menores que o k-ésimo elemento são movidos antes desse elemento e todos iguais ou maiores são movidos para trás dele. A ordenação dos elementos nas duas partições é indefinida. Ela retorna uma matriz de índices da mesma forma que arr, ou seja, arr[index_array]gera uma partição de arr.

```python
# função par manter lista de idx que devem ser mantidos em ordem ordenado, é usada para realizar uma partição indireta.
def indices_of_top_k(arr, k):
    return np.sort(np.argpartition(np.array(arr), -k)[-k:])

class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k
        
    def fit(self, X, y=None):
        self.feature_indices_ = indices_of_top_k(self.feature_importances, self.k)
        return self
    
    def transform(self, X):
        return X[:, self.feature_indices_]
```

Se usarmos o KFold para um conjunto de dados desbalanceado, podemos acabar  com dados de treinamento que contém poucas ou nenhuma classe minoritária.Portanto, para evitar esse problema, usamos o StratifiedKFold. **StratifiedKFold é uma versão melhorada de k-Fold que retorna folds estratificadas: cada conjunto contém aproximadamente a mesma porcentagem de amostras de cada classe de destino que o conjunto completo**.

Em outras palavras, StratifiedKFold embaralha seus dados (amostras sem reposição), depois divide os dados em n_splits partes e pronto. Agora, ele usará cada parte como um conjunto de teste. Observe que apenas e sempre embaralha os dados uma vez antes de dividir. Isso significa que cada parte preserva a mesma porcentagem de amostras de cada classe (rótulo) que os dados originais.

>>> Objetivo da **regularização** é reduzir o problemas de overfitting, para não ter o modelo tendencioso e ineficiente, ou seja, para usa enfim os melhores recursos do conjunto de dados

Essa é uma forma de executar automaticamente a seleção dos  recuros (variável), o que é bom.**À medida que seu valor aumentar, menor peso será dado aos atributos que são menos importantes.**

### 7.5.1. O que são Hiperparâmetros?

<font color="red">Apesar de a palavra hiperparâmetros parecer um bicho de sete cabeças, ela  simplesmente se refere a números que você usa para acertar ou ajustar o  desempenho de um modelo estatístico ou de aprendizado de máquina.Quando se  trata de otimizar um modelo, você testa diversas variações de hiperparâmetros  nesse modelo, geralmente recorrendo ao grid search para identificar esses  parâmetros. Ou seja, você constrói um “grid” de todas as combinações possíveis  de hiperparâmetros em um espaço de busca e experimenta todas.(Aileen Nielsen, 2021)</font>

**Teoricamenteo um  hiperparâmetro é um
parâmetro de um algoritmo de aprendizado (não do modelo). Como tal, não
é afetado pelo próprio algoritmo de aprendizado; deve ser definido antes do
treinamento e permanecer constante ao longo dele.**


Nessa mesma linha de trabalho **Grid Search** um dos modos de fazer isso seria manipular manualmente os hiperparâmetros até identificar uma **ótima combinação de valores**. Este seria um trabalho muito **tedioso**, e talvez você 
não tenha tempo para explorar muitas combinações.Já por outro lado se estiver explorando relativamente poucas combinações, a abordagem da **Grid Search** é um boa opção, mas quando o espaço de busca do hiperparâmetro for grande, é preferível utilizar **Randomized Search** em seu lugar. Esta método pode utilizar as classes (GridSearchCV ou RandomizedSearchCV do Scikit-Learn, para efetuar a busca)  , **mas, em vez de tentar todas as 
combinações possíveis, ela seleciona um valor aleatório para cada hiperparâmetro em  cada iteração e avalia um determinado número de combinações aleatórias**

Em síntese ,o **Grid Search** realiza metodicamente combinações entre todos os parâmetros do algoritmo,criando um grande grid com todos as combinações possíveis.O **Randomized Search** gera amostras dos parâmetros dos algoritmos a partir de uma distribuição randômica uniforme para um número fixo de interações. Um modelo é construído e testado para cada combinação de parâmetros.

### 9. Classes desbalanceadas

Segundo (CHAN, 2019) em aprendizado de máquina, trabalhar com classes **desbalanceadas** pode ser um desafio. Classes desbalanceadas ocorrem quando há uma representação desigual de classes, onde a classe minoritária (será explicado nos capítulo posteriomente) **é muitas vezes de maior interesse para nós analisamos**. Isso é proeminente em áreas onde a coleta de dados das classes minoritárias é difícil devido a restrições de tempo e custo. Por exemplo, se você considerar problemas de classificação como não-fraude versus fraude e não-câncer versus câncer, transações fraudulentas e pacientes com câncer são mais raras do que transações normais e pacientes saudáveis. Outros exemplos incluem detecção de derramamento de óleo, detecção de intrusão de rede, outras doenças raras, etc.

### 9.1. Desafios

Observa-se que (CHAN, 2019),<font color="red"> afirma que os algoritmos de aprendizado de máquina são criados **para minimizar erros</font>**. Para conjuntos de dados não balanceados, como as instâncias da classe majoritária  estão superando as classes minoritárias, os classificadores provavelmente classificarão novos dados para a classe majoritária.

<font color="coulor">Além disso, (CHAN, 2019) simplifica esse conceito por meio de um exemplo, se você tiver um conjunto de dados com 90% de não câncer (saudável) e 10% de câncer, o classificador será tendencioso para classificar uma nova pessoa como sem câncer (saudável), pois seria 90% preciso. Isso leva a False Negatives, que no caso do câncer, pode ser visto como mais caro do que ter False Positives! Diz-se a um paciente que não tem câncer, quando na verdade tem... você vê o problema aí?</font>

### 9.2. O Problema da classe Rara

Segundo (ANDREW e BRUCE, 2019) afirmam em muitos casos, existe um **desequilíbrio nas classes** a serem previstas, com uma classe muito mais **prevalente que a outra** por exemplo, pedidos de seguro legítimos versus fraudulentos, ou curiosos versus compradores em um  site. A classe rara (por exemplo, os pedidos fraudulentos) geralmente é a  classe de maior interesse, e costuma ser designada como 1, ao contrário dos  mais prevalentes 0s. Em um cenário típico, os 1s são o caso mais importante,  no sentido de que classificá-los erroneamente como 0s fica mais caro do que  classificar erroneamente 0s como 1s. Por exemplo, identificar corretamente  um pedido de seguro fraudulento pode economizar milhares de dólares. Por  outro lado, identificar corretamente um pedido não fraudulento economiza  apenas o custo e o tempo de passar o pedido manualmente com uma revisão  mais cuidadosa (que é o que você faria se o pedido fosse rotulado como “fraudulento”).

**Nesses casos, a menos que as classes sejam facilmente separáveis, o modelo de classificação mais preciso pode ser aquele que simplesmente classifica tudo como 0. Por exemplo, se apenas 0,1% dos visitantes em uma loja online faz uma compra, um modelo que prevê que cada visitante sairá sem fazer uma compra será 99,9% preciso. No entanto, será inútil. Em vez disso, seria melhor ter um modelo que seja menos preciso no geral, mas bom em identificar os compradores, mesmo que classifique erroneamente alguns não compradores no meio do caminho.**


><font color="red">Em estatística, é comum na prática  atribuir “1s” ao resultado mais raro, ou seja, são o caso mais importante, conhecido **Falso positivo**. Por outro lado, é comum atribuir "0s" ao resultado mais prevalente, conhecido **Falso Negativo</font>**.

Bem como, as classes desbalanceadas são tendenciosas e os modelos de aprendizado de máquina tendem a prever (mais) a classe majoritária. Você não quer que seu modelo preditivo ignore as classes minoritárias, certo? É por isso que existem técnicas para superar o problema do desbalanceamento; **sub-amostragem e super-amostragem**. Qual a diferença entre essas duas tecnologias?


- **<font color="red">Undersampling:</font>** A sub-amostragem diminuiria a proporção de sua classe majoritária até que o número seja semelhante ao da classe minoritária ao mesmo tempo.

- **<font color="red">Oversampling:</font>** A super-amostragem iria reamostrar a proporção da classe minoritária seguindo a proporção da classe majoritária.




(ESCOVEDO, 2020) ilustra esse conceito por meio da figura acima do autor (ALENCAR) <a href="https://www.kaggle.com/code/rafjaa/resampling-strategies-for-imbalanced-datasets/notebook">Resampling</a>
    
    
Assim temos:

- Undersampling - Duplicando amostras da classe minoritária
- Oversampling - Excluindo amostras da classe majoritária.

Na maioria dos casos, a **super-amostragem** é mais preferível às vezes ao invés de técnicas de **sub-amostragem**. O motivo é que, na **sub-amostragem**, tendemos a remover instâncias de dados que podem conter algumas informações importantes, por outro  na **super-amostragem** lado pode ser mais trabalhoso e mais complicado de explicar, mas aumenta o tamanho do conjunto de dados criando dados sintéticos baseados em regras estatísticas e aleatoriamente,

Segundo (ESCOVEDO, 2020) apura que a **acurácia** é a métrica mais comum para problemas de classificação, embora nem sempre seja adequada, sendo potencialmente perigosa quando usada em datasets desbalanceados.Nessa mesma linha de raciocínio, se **86%** dos dados referente aos clientes que **não efeturam churn** pertence à classe negativa (**Cancelamento -> NO**), é fácil ver que um **classificador dummy** (variáveis dummy são variáveis binárias (0 ou 1)) terá acurácia entorno de **85% á 90% precisa**, simplesmente prevendo a <font color="red">classe negativa sempre</font>, o que não é aceitável na prática, pois não estaremos identificando os exemplos pertencentes à **classe positiva**. Para tal, podemos usar métricas alternativas como **precisão (precision)** e **revocação (recall**) que permitem que o desempenho do modelo seja considerado com foco na classe minoritária (neste caso, **se cliente efetuará churn (Sair -> Sim)**.Dessa forma, para entender estas métricas, é importante definirmos os conceitos de Verdadeiro Positivo, Falso Positivo, Verdadeiro Negativo e Falso Negativo:


- **Verdadeiro Positivo (VP ou TP):** a classe verdadeira é positivo, e a saída do modelo é positiva (acertou);

- **Falso Positivo (FP):** a classe verdadeira é negativo, mas a saída do modelo é positiva (errou);

- **Verdadeiro Negativo (VN ou TN):** a classe verdadeira é negativo, e a saída do modelo é negativa (acertou);

- **Falso Negativo (FN):** a classe verdadeira é positivo, mas a saída do modelo é negativa (errou).

Assim sendo, **<font color="red">as instâncias positivas e negativas que um classificador prevê corretamente são chamadas verdadeiros positivos (VP ou TP) e  verdadeiros negativos (VN ou TN), respectivamente</font>**.**<font color="blue"> Da mesma forma, as instâncias classificadas incorretamente são chamadas de falsos positivos (FP) e falsos negativos (FN)</font>**.

Temos mais alguns métricas ainda  a ser consideradas que são explicadas por (ESCOVEDO, 2020) da seguinte maneira: 

### 9.3. Precision 

A **precision** calcula a proporção do número total de exemplos classificados como positivos corretamente (verdadeiros positivos) dividido pelo número total de exemplos classificados como positivo pelo modelo (corretamente ou incorretamente). Quando maximizamos a precisão (mais próximo de 1), ou seja, **é acurácia das predições positivas**, minimizamos os **falsos positivos**:


**Precision** tentar responder a seguinte pergunta?


    Qual é a proporção de identificações (classificações) positivas estava realmente correta?


$$Precision = \frac{VP}{(VP + FP)}$$  

**<center> outra abreviação, que você pode encontra em literaturas  </center>**


$$Precision = \frac{TP}{(TP + FP)}$$

> **<font color="red">A precision é a métrica que  dá a informação de quantas observações o modelo classificou corretamente como 1s</font>**



>Em suma, **precision**(o percentual de classificações previstas que estão corretas) é apenas o primeiro passo na avaliação de um modelo.


### 9.4. Recall ou Sensitivity

A **recall** calcula a proporção do número total de exemplos classificados como positivos corretamente (verdadeiros positivos) dividido pelo número total de exemplos com classe positiva. Quando maximizamos o recall, (mais próximo de 1), minimizamos os **falsos negativos**.

$$ Recall = \frac {VP} {(VP + FN)}$$

$$ Recall = \frac {TP} {(TP + FN)}$$

>**<font color="red">A recall, também conhecida como sensitivity ou taxa de verdadeiros positivos (TPR), mede a força do  modelo em prever um resultado positivo — a proporção de 1s que identifica corretamente.</font>**

Na prática, quase sempre temos que escolher entre uma alta **precision** ou um alto **recall** pois, tipicamente, é impossível ter ambos um trade-off.De acordo com problema, possivelmente priorizaríamos ter uma alta **precision** (minimizar a quantidade do número de clientes com potencial de efetuar **Churn** não detectados) e podemos tolerar um recall mais **baixo** (alguns clientes **sem** potencial de cancelar a sua **assinatura**, mas detectados como potencial de cancelamento) (ESCOVEDO, 2020).

Com efeito pronunciado, **precision e recall** são usadas para avaliar o desempenho de sistema de classificação (modelo).Em poucas palavras, dá informações sobre a qualidade do modelo em estimar a classe de interesse.

Para avaliar completamente a eficácia de um modelo, você deve examinar tanto a **precision** quanto a **recall**. Infelizmente, **precision e recall** estão frequentemente em tensão.**<font color="red">Ou seja, em geral, há um compromisso: à medida a recall aumenta, a precision diminui – e vice-versa.Isso se chama trade-off de precision/recall</font>**


### 9.5. Specificity

Em poucas palavras, a specificity estima (mede) a capacidade do modelo em preve um  resultado **negativo**.Logo, temos a seguinte equação:


$$ specificity =  \frac{TN}{TN+FN}$$

A  Specificity também usa TN, o número de verdadeiros negativos. Isso significa que a Sensitivity e Specificity usam todos os quatro números da confusion matriz, em oposição à **precision e recall**, que usam apenas três.

### 9.6. F-Score ou F1


Normalmente, as pontuações da **precision e recall** são dadas juntas e não são citadas individualmente. Isso ocorre porque é fácil variar a sensibilidade de um modelo para melhorar a **precision** em desfavor  **recall**, ou vice-versa (WOOD).

O desempenho de um modelo pode ser resumido por uma única pontuação que calcula a média da **precision e recall**, denominada **F-Score ou F1**. A maximização da F-Score maximizará a **precison e a recall** ao mesmo tempo, conforme ilustra a figura em seguida:



> Image source (CHAN, 2019)

A F-Score  é a média <a href="https://deepai.org/machine-learning-glossary-and-terms/harmonic-mean">harmônica</a> da **precision e recall** . Ao passo que a média regular trata igualmente todos os valores, a média harmônica dá mais importância aos valores mais baixos. Como resultado, o classificador só obterá um score F-Score alto se **precision e  recall** forem altas (valor mais alto possível de uma F-score é 1, indicando **precision e recall** perfeitas).A F-Score favorece os classificadores que têm **precision e recall** semelhantes e F-score é maneira simples de **comparar dois ou  classificadores**. A expressão é dada pela  seguinte equação

$$ F1 = 2 x​​\frac{precision​​x​​recall}{precision + recall}$$

Assim de tal modo que  nos permite combinar a **precision e  recall** em um único número.

Acrescente-se que,se considerarmos a **precision e recall** mais importantes que a outra, podemos usar o escore F β , que é uma média harmônica ponderada de **precision e recall**. Isso é útil, por exemplo, no caso de um teste médico, onde um falso negativo pode ser extremamente caro em comparação com um falso positivo. ### 9.7. Calculando Precision e Recall vs F-Score

Observa-se que (WOOD) refina o exemplo citado anterior **diagnóstico de doença**, obtivemos **precision de 0,625 e recall de 0,56**.

Imagine que consideramos **precicion e recall** como sendo de igual importância para nossos propósitos. Nesse caso, usaremos o F-score para resumir a **precision e o recall** juntas.

Colocando os números da **precision e recall** na fórmula para o F-score, obtemos:

$$ F1 = 2 x​​\frac{ precision​x​recall} { precision + recall} $$

$$ F1 = 2 x ​​\frac {0.625 * 0.56} {0.625 + 0.56} = 0.59 $$


Observe que o F-score de **0,59** está entre os valores de precision e recall **(0,625 e 0,56)**. Dessa maneira, isso ilustra como o F-score pode ser uma maneira conveniente de calcular a média da **precison  e recall** para combinar em um único número.


### 9.8. Calculando Precision e Recall vs Sensitivity e Specificity 

Podemos calcular **sensitivity e specificity** para o caso referente ao exemplo citado anterior **diagnóstico de doença**

Assim temos; TP= 5, FP= 3 e TN= 8, lembre-se,esses valores são da confusion matrix;

Detacando-se, **sensitivity** é claro sai como o mesmo valor que a **recall (na classificação binária, bioestatística e diagnósticos médicos);**

$$ sensitivity = \frac {TP} {(TP + FN)} =  \frac {5} {(5 + 4)} = 0.56$$

Considerando a **specificity** temos;

$$ specificity =  \frac{TN}{TN+FN} = \frac{8}{8+3} = 0.73$$


O número de TN (verdadeiros negativos) corresponde ao número de pacientes identificados pelo teste como portadores da **doença quando não tinham a doença**

Assim, usando uma interpretação probabilística, podemos ver a **specificity** como a probabilidade de um teste **negativo** dado que o paciente está bem, enquanto a **sensitivity** é a probabilidade de um teste **positivo** dado que o paciente **tem a doença**.

Portanto, **sensitivity** e **specificity** são preferidas ao invés da **precision  e recall** na área de domínio da medicina, enquanto **precision recall** são as métricas mais comumente usadas para recuperação de informações.**<font color="red">Porém, isso inicialmente parece estranho, já que ambas  técnicas de métricas estão medindo a mesma coisa: o desempenho de um classificador binário</font>**.


Constituir, a razão para essa **divergência** é que, quando estamos medindo o desempenho de um classificador de pesquisa, nos preocupamos apenas com os resultados retornados, portanto, tanto a **precision** quanto a **recall** são medidos em termos de **verdadeiros positivos(TP) e falsos positivos (FP)**. No entanto, se estivermos testando um classificador específico voltado para área da medicina, é importante levar em consideração o número de **verdadeiros negativos (TN)**, pois estes representam o grande número de pacientes que **não possuem a doença** e foram categorizados corretamente pelo classificador.

### 9.8. ROC curve 

De acordo (GOOGLE DEVELOPERS) uma curve ROC ( curva característica de operação do receptor, em português) é outra ferramenta comum usada com classificadores binários.É gráfico que mostra o desempenho de um modelo de classificação em todos os limites de classificação. Esta curva traça dois parâmetros:

- Taxa de Verdadeiro Positivos (True Positive Rate)
- Taxa de Falsos Positivos (False Positive Rate)


True Positive Rate ( TPR ) é sinônimo da **recall** e, portanto, é definido da seguinte forma:

$$ TPR = \frac{TP}{TP+FN}$$

False Positive Rate (FPR) é definida da seguinte forma:

$$ FPR = \frac{FP}{FP+TN}$$
>**É a proporção de instâncias negativas classificadas incorretamente como
positivas.**


Uma ROC curve plota TPR vs FPR em diferentes limites de classificação. A redução do limite de classificação, classifica mais itens como positivos, aumentando assim tanto os falsos positivos quanto os verdadeiros positivos, ou seja, mais uma vez existe um trade-off quanto maior a recall (TPR), mais falsos positivos (FPR) o classificador produz. A figura a seguir mostra uma curva ROC típica (GOOGLE DEVELOPERS).

O processo ilustrado na figura segue o seguinte passo:

2. Calcule a **specificity  e a recall** cumulativas com base nas amostras (registros) organizadas.

### 9.9. AUC

De acordo (GOOGLE DEVELOPERS) AUC significa "Área sob a Curva ROC". Ou seja, a AUC mede toda a área bidimensional abaixo de toda a curva ROC (pense no cálculo integral) de (0,0) a (1,1).Em outras palavras, <font color="re">representa o grau ou medida de separabilidade. AUC informa o quanto o modelo é capaz de distinguir entre as classes. Quanto maior a AUC, melhor o modelo é para prever **0s** classes como 0 e **1s** classes como 1.

Então, o que é SMOTE? A técnica SMOTE ou Synthetic Minority Oversampling é uma técnica de super -amostragem de minoria sintética, mas o SMOTE funciona de maneira diferente da super-amostragem típica, ou seja, SMOTE é usado para sintetizar dados onde os recursos são contínuos,em um problema de classificação.

**É uma técnica de super-amostragem onde as amostras sintéticas são geradas para a classe minoritária. Este algoritmo ajuda a superar o problema de overfitting colocado pela super-amostragem aleatória. Ele se concentra no espaço de recursos para gerar novas instâncias com a ajuda da interpolação entre as instâncias positivas que estão juntas. ### 10.1. Balaceamento de classe ADASYN: Adaptive Synthetic Sampling Approach

ADASYN é uma forma generalizada do algoritmo **<font color="red">SMOTE</font>**. Esse algoritmo também visa a super- amostrar a classe minoritária, gerando instâncias sintéticas para ela. Mas a diferença aqui é que considera a distribuição de densidade, ele que decide o não. de instâncias sintéticas geradas para amostras difíceis de aprender. Devido a isso, ajuda a mudar de forma adaptativa os limites de decisão com base nas amostras difíceis de aprender. Esta é a principal diferença em relação ao SMOTE.(SWASTIK SATPATHY, 2020)

--- 
Amostragem sintética adaptativa (ADASYN) aprender a partir de conjuntos de dados desbalanceados. A ideia essencial do ADASYN é usar uma distribuição ponderada para diferentes exemplos de classes minoritárias de acordo com seu nível de dificuldade de aprendizado, onde mais dados sintéticos são gerados para exemplos de classes minoritárias que são mais difíceis de aprender em comparação com os exemplos minoritários que são mais fáceis de aprender. Como resultado, a abordagem ADASYN melhora o aprendizado em relação às distribuições de dados de duas maneiras: 

- (1) reduzindo o viés introduzido pelo desequilíbrio de classe

- (2) mudando adaptativamente o limite de decisão de classificação para os exemplos difíceis. 

A geração de dados sintéticos seria inversamente proporcional à densidade da classe minoritária. Isso significa que mais dados sintéticos são criados em regiões do espaço de recursos onde a densidade de exemplos minoritários é baixa e menos ou nenhum onde a densidade é alta.Em termos mais simples, em uma área onde a classe minoritária é menos densa, os dados sintéticos são criados mais. Caso contrário, os dados sintéticos não são feitos tanto.


Os problemas podem estar nos outliers. Assim como, ADASYN se concentraria nos dados de densidade onde a densidade é baixa. Freqüentemente, os dados de baixa densidade são discrepantes. A abordagem ADASYN então colocaria muita atenção nessas áreas do espaço de recursos, o que pode resultar em pior desempenho do modelo. Pode ser melhor remover o outlier antes de usar o ADASYN.

### 11. A maldição da dimensionlidade

Segundo (GÉRON, 2021) apura que  muitos problemas de aprendizado de **máquina envolvem milhares ou até milhões de características para cada instância de treinamento**. Todas essas características não somente tornam o treinamento extremamente lento, como também dificultam e muito encontrar uma boa solução, conforme veremos.**<font color="red"> Esse problema costuma ser chamado de a maldição da dimensionalidade</font>**.

Em outras palavras, a maldição da dimensionalidade refere-se ao fato de que muitos problemas que não existem em um espaço de dimensão inferior surgem no espaço de alta dimensão. Uma manifestão comum no aprendizado de máquina é o fato de os vetores de alta dimensão amostrados aleatoriamente geralmente serem muito esparsos, aumentando o risco de sobreajuste e dificultando bastante a identifcação de padrões nos dados sem ter muitos dados de treinamentos, (GÉRON, 2021).

Apesar de estamos tão habituados a viver em **três dimensões** que nossa lógica nos trai usando tentamos imaginar um espaço de alta dimensão. Nossa mente tem uma dificuldade enorme de imaginar um mero hipercubo 4D básico (veja aFigura adiante), que dirá um elipsoide de 200 dimensões dobrado em um espaço de mil dimensões (GÉRON, 2021).



> Image source: (GÉRON, 2021)


**Em teoria, aprimora uma solução para a maldição da dimensionalidade talvez fosse aumentar o tamanho do conjunto de treinamento a fim de alcançar uma densidade suficiente de instâncias de treinamento. Mas infelizmente, na prática, o número de instâncias de treinamento necessárias para alcançar uma determinada densidade cresce exponencialmente com o número de dimensões** (GÉRON, 2021).

**As principais motivações para usa redução da dimensionalidade são:**

   + Acelera um algoritmo de treinamento subsequente ( em algumns casos, ele pode até mesmo remover ruídos e características redundantes, fazendo com que o algoritmo de treinamento tenha um melhor desempenho;
    
   + Visualizar os dados  e obter insights sobre as caraterística mais importantes;
    
   + Simplesmente economizar espaço (compressão)
    
**As principais desvantagens são:**
  + Algumas informações são perdidas, possivelmente degradando o desempenho de algoritmos de treinamentos subsquentes;
    
  + Poder ser intensivo  computacionalmente;
  
  + Adicona algumas complexidade aos seus pipelines de Aprendizado de Máquina
  
  + Característica (variáveis) transformadas geralmente são difícil a interpretação.

### 11.1.2. Aprendizado manifold

De acordo com (GÉRON,2021) O conjunto Swiss roll é um exemplo de manifold 2D. Em termos simples, um manifold 2D é uma forma 2D que pode ser dobrada e retorcida em um espaço dimensional maior. De modo mais geral, um manifold d-dimensional é uma parte de um espaço n-dimensional (sendo d < n) que se assemelha localmente a um hiperplano d-dimensional. No caso do Swiss roll, d = 2 e n = 3: assemelha-se localmente a um plano 2D, mas é enrolado na terceira dimensão.

Observa-se (GÉRON, 2021) apura que muitos algoritmos de redução de dimensionalidade funcionam modelando o manifold em que estão as instâncias de treinamento; isso se chama de **aprendizado manifold**. Ele se baseia no **manifold assumption**, também chamada de <font color="red">hipóteses múltiplas</font>, que sustenta que a maioria dos conjuntos de dados de alta dimensão do mundo real fica próxima a um manifold de dimensões bem mais baixas. Essa suposição é muitas vezes estudada empiricamente.

Em suma, reduzir a dimensionalidade do seu conjunto de treinamento antes do treinamento de um modelo geralmente acelera o treinamento, mas nem sempre resulta em uma solução melhor ou mais simples; tudo depende do conjunto de dados.**Diante do exposto, abordaremos um dos algoritmos mais popular de redução dimensionalidade o PCA.**

### 11.2. Análise de Componentes Principais (PCA)

**O PCA foi inventado em 1901 por Karl Pearson e utiliza álgebra linear para transformar datasets em uma forma comprimida, o que é geralmente conhecido como Redução de Dimensionalidade. Com PCA você pode escolher o número de dimensões (chamados componentes principais) no resultado transformado.**

O PCA é um método para extração das variáveis importantes (na forma de componentes) a partir de um grande conjunto de variáveis, disponíveis em um conjunto de dados. Esta técnica permite extrair um número pequenos de conjuntos dimensionais a partir de um dataset altamente dimensional. Com menos variáveis a visualização também se torna muito mais significativa. PCA é mais útil quando se lida com 3 ou mais dimensões.

![pca.png](http://www.nlpca.org/fig_pca_principal_component_analysis.png)

> Image source: http://www.nlpca.org/fig_pca_principal_component_analysis.png

Cada componente resultante é uma combinação linear de n atributos. Ou seja, cada componente principal é uma combinação de atributos presentes no dataset. O Primeiro **Componente Principal** é a combinação linear dos atributos com máxima variância e determina a direção em que há mais alta variabilidade nos dados. Quanto maior a variabilidade capturada no primeiro componente principal, mais informação será capturada pelo componente. O Segundo **Componente Principal** captura a variabilidade remanescente. Todos os componentes subsequentes possuem o mesmo conceito.

**As principais vantanges de usar o PCA, segundo (SALAS):**

- 2°: Reduzir o problema de esparsidade de dados.

- Os Eigenvectors (Autovetores) determinam as direções do novo espaço de recursos.

<font color="red"> O PCA não tem desvantagens. Se for usado de forma corretamente, ele deve filtrar o ruído e você deve ficar com um sinal mais forte.</font>

No entanto, uma questão relacionada é: quais são as limitações do PCA, ou seja, quando você pode aplicar o PCA em um conjunto de dados para redução de dimensionalidade. Para isso, precisamos conhecer as premissas que são utilizadas no PCA:

- **Linearidade:** O PCA assume que os componentes principais são uma combinação linear dos recursos originais. Se isso não for verdade, o PCA não fornecerá resultados sensatos.

- **Grande variância implica mais estrutura:** o PCA usa variância como medida de quão importante é uma dimensão específica. Assim, eixos de alta variância são tratados como componentes principais, enquanto eixos de baixa variância são tratados como ruído.

- **Ortogonalidade:** PCA assume que os componentes principais são ortogonais.

De acordo com os autores (ANDREW e BRUCE, 2019) dizem que normalmente as variáveis variam em conjunto **(covariam)**, e algumas variações em uma delas são, na verdade, duplicadas pela variação em outra. A análise de componentes principais (PCA) é uma técnica para descobrir o modo como as variáveis numéricas covariam.
Alguns Termos-chave para o PCA destacado por (ANDREW e BRUCE, 2019): 

Componente principal

- **Uma combinação linear das variáveis preditoras.** 

Cargas 

- **Os pesos que transformam as preditoras em componentes.** 
- **Sinônimo: pesos**


<font color="red">A ideia na PCA é combinar múltiplas variáveis preditoras numéricas em um conjunto menor de variáveis, que são combinações lineares ponderadas do conjunto original. O menor conjunto de variáveis, os componentes principais, “explica” a maior parte da variabilidade do conjunto completo de variáveis, reduzindo a dimensão dos dados. Os pesos usados para formar os componentes principais revelam as contribuições relativas das variáveis  originais para os novos componentes principais (ANDREW e BRUCE, 2019)</font>. 


De fato, nesse mesmo sentido (ANDREW e BRUCE, 2019) explica que o  PCA foi originalmente da proposta por **Karl Pearson**. No que foi possivelmente o primeiro artigo sobre aprendizado não supervisionado, **Pearson** reconheceu que em muitos problemas existe variabilidade **(variabilidade, também chamada de dispersão, mede se os valores de dados estão compactados ou espalhados)** nas variáveis preditoras; então, ele desenvolveu a PCA como uma técnica para modelar essa variabilidade.


**Calculando os componentes principais**

Além disso, (ANDREW E BRUCE, 2019) acrescente-se que passar de duas variáveis para mais variáveis é fácil. Como resultado, para o primeiro componente, simplesmente inclua as variáveis preditoras adicionais na combinação linear, atribuindo pesos que otimizem a coleção da covariação de todas as variáveis preditoras nesse primeiro componente principal (covariância).**<font color="red">O cálculo dos componentes principais é um método estatístico clássico, baseado na matriz de correlação dos dados ou na matriz de covariância, e é executado rapidamente, sem depender de iterações.Porém,funciona apenas com variáveis numéricas, não com  categóricas.</font>**

### 11.2.1. Preservando a variância

De acordo (GÉRON, 2021) ressalta que antes de projetar o conjunto de treinamento em um hiperplano de menor dimensão, primeiro é necessário escolher o hiperplano correto. Por exemplo, um conjunto de dados 2D simples é representado à esquerda na Figura adiante, junto com três eixos diferentes (ou seja, hiperplanos 1D). À direita, está o resultado da projeção do conjunto de dados em cada um desses eixos. Como pode observar, a projeção na linha contínua preserva a variância máxima, enquanto a projeção na linha pontilhada preserva muito
pouco da variância e a projeção na linha tracejada preserva uma quantidade intermediária de variância.



> Image source: (GÉRON, 2021)


Ness sentido, parece aceitável selecionar o eixo que preserva a quantidade máxima de variância, porque é bem provável que ela perca menos informações do que as outras projeções. Outra forma de justificar essa escolha é que o eixo minimiza a distância quadrada média entre o conjunto de dados original e sua projeção nesse eixo. Essa é a ideia bastante simples por trás da PCA (GÉRON, 2021).

PCA identifica o eixo que responde pela maior quantidade de variânciano conjunto de treinamento. Na Figura mencionada anteriomente, é a linha contínua. Ela também identifica um segundo eixo, ortogonal ao primeiro, responsável pela **maior variância remanescente**. Nesse exemplo 2D, não existe escolha: é a linha pontilhada. Se fosse um conjunto de dados de dimensão maior, a PCA também encontraria um terceiro eixo, ortogonal aos dois
eixos anteriores, e um quarto, um quinto e assim sucessivamente identificaria tantos eixos quanto o número de dimensões no conjunto de dados.


O *i-ésimo* eixo se chama *i-ésimo* componente principal (PC) dos dados. Na Figura ilustra anterior, o primeiro PC é o eixo no qual o vetor c1 se encontra, e o segundo PC é o eixo no qual o vetor c2 se localiza. Na Figura ilustra no capítulo (10.1.1.) , os dois primeiros PCs são os eixos **ortogonais nos quais as duas setas se encontram no plano, e o terceiro PC é o eixo ortogonal a esse plano**.

### 12.1. Acurácia (accuracy)

É a porcentagem de classificações corretas, mede a precisão (é a fração de previsões que nosso modelo acertou.) do modelo de classificação, como a proporção de  resultados verdadeiros em relação ao total de casos analisados. Quanto maior, melhor!

Assim, temos a expressão que é calculada da seguinte forma com os os termos são positivos e negativos;;


$$ Accuracy = \frac {\sum {TP} + \sum  {TN}}  {SampleSize} $$ 


ou 
$$ Accuracy = \frac {Número​​total​​de​​previsões​​corretas} {Total​​previsões } $$ 


Considere a expressão que é calculada para **classificação binária** 



$$ Accuracy =\frac {(TP + TN)} {(TP + TN + FP + FN)}$$

Onde:

- TP = Verdadeiros Positivos
- TN = Verdadeiros Negativos
- FP= Falso Positivo
- FN = Falso Negativo

**O que é uma boa acurácia?** Isso depende. Se estou fazendo a predição de uma fraude (que, em geral, é um evento raro, por exemplo, 1 em 10 mil), eu poderia ter uma acurácia muito elevada sempre prevendo que não haveria uma fraude. Porém, esse modelo não seria muito útil. <font color="red"> A accuracy é simplesmente uma medida do erro total </font>


Assim, acrescenta-se que  modelo atinge uma acurracy em teste
**95% (proporção de predições corretas)**, isso é bom, mas nem tudo é perfeito assim, **ressaltando o problema problema da classe rara, não é recomendado usar a acurácia para classes desbalanceadas**.

### 12.2. Discrimination threshold

A maioria dos classificadores binários que fazem a predição de probabilidades tem um **discrimination threshold** de **50%**. Se a probabilidade prevista estiver acima de **50%**, o classificador atribuirá um rótulo positivo.

Desde modo, como regra geral, um **threshold** é equilibrado entre os casos é definido de **0,5 ou 50% de probabilidade**. No entanto, esse **threshold** pode não ser o **threshold** ideal: muitas vezes há uma relação inversa entre **precision e recall** em relação a um **Discrimination threshold**.Logo, ajustando o **threshold** do classificador, é possível ajustar a pontuação de F1 (a média harmônica da precision e recall) ao melhor ajuste possível ou ajustar o classificador para se comportar de forma ideal para a aplicação específica.

```python
# Definindo uma função, para realizar o cálculo do Intervalo de Confiança, da acurácia e do erro de modelos 
# preditivos de classificação.

def CI(score, nElements,  ci = .95):

    # Define o Z-score, a ser utilizado para o Intervalo de Confiança selecionado.
    
    z = {
        .90  : 1.645,      # Para Intervalos com 90% de Confiança.
        .95  : 1.96,       # Para Intervalos com 95% de Confiança.
        .98  : 2.326,      # Para Intervalos com 98% de Confiança.
        .99  : 2.576,      # Para Intervalos com 99% de Confiança.
        .995 : 2.807,      # Para Intervalos com 99.5% de Confiança.
        .999 : 3.291       # Para Intervalos com 99.9% de Confiança.
    }

    # Calculando o range de variação do Intervalo.

    interval = z.get(ci) * np.sqrt( (score * (1 - score)) / nElements)

    # Retornando do Intervalo de Confiança obtido.

    return score - interval, score + interval
```

- **Inicialização: Lecun_normal**
- **Otimizador: Nadam**
- **Learning Rate: Performance Scheduling**
- **Regularização: AlphaDropout**


```python
# Limpamos a sessão Keras e gerando semente aleatória
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# create model
model_ada = keras.models.Sequential([
    keras.layers.Dense(25,input_dim=10,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.AlphaDropout(rate=0.25),
    keras.layers.Dense(1,  activation="sigmoid")
])

# Compilando modelo 
model_ada.compile(loss='binary_crossentropy',
              optimizer = keras.optimizers.Adam(lr=0.003),
               metrics=['accuracy','Precision','Recall','AUC'])


"""
salva os pontos de verificação do modelo em intervalos regulares durante o treinamento, por padrão no fim de cada época:
"""
model_checkpoint_cb = keras.callbacks.ModelCheckpoint("models/churn_model_ada_selu.h5", save_best_only=True)


"""
Uma forma de implementar a early stopping é simplesmente usar a callback da EarlyStopping. Ela interromperá o treinamento quando não calcular
nenhum progresso na validação definida para várias épocas (definida pelo argumento patience) e, opcionalmente, reverterá para o melhor modelo. Você
pode combinar as duas callbacks para salvar os pontos de verificação do modelo (no caso de o computador travar) e interromper o treinamento com
antecedência, quando não houver mais progresso (a fim de evitar perda de tempo e de recursos)
"""
early_stopping_cb = keras.callbacks.EarlyStopping(patience=20)

#callback TensorBoard()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)

# cria um lista de callbacks
callbacks = [early_stopping_cb,model_checkpoint_cb, tensorboard_cb,TqdmCallback(verbose=0)]

# Treina modelo
model_ada.fit(X_train_min_max_ada, y_train_ada, epochs=60,batch_size=20,
                              validation_split=0.1,
                              verbose=0,
                              callbacks=callbacks)

# Carrega e Avalia o modelo
model_ada = keras.models.load_model("models/churn_model_ada_selu.h5") # retornar para o melhor modelo
```

```python
# Limpamos a sessão Keras e gerando semente aleatória
keras.backend.clear_session()
tf.random.set_seed(42)
np.random.seed(42)

# create model
model_blsm = keras.models.Sequential([
    keras.layers.Dense(25, input_dim=10,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.Dense(25,activation="selu",kernel_initializer="lecun_normal"),
    keras.layers.AlphaDropout(rate=0.1),
    keras.layers.Dense(1,  activation="sigmoid")
])

# Compilando modelo 
model_blsm.compile(loss='binary_crossentropy',
              optimizer = keras.optimizers.Nadam(lr=0.004),
              metrics=['accuracy','Precision','Recall','AUC'])




""""
Na linha do código a seguir, ela calcule o erro de validação a cada N etapas (exatamente como na early stopping) e reduza a taxa de aprendizado por um fator de λ quando o erro parar de diminuir.
A callback ReduceLROnPlateau, ela multiplicará a taxa de aprendizado por 0,5  sempre que a melhor perda de validação não melhorar por 5 épocas consecutivas

"""
lr_scheduler = keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=10)
model_checkpoint_cb = keras.callbacks.ModelCheckpoint("models/churn_model_blsm_selu_nadam.h5", save_best_only=True)
run_logdir2 = get_run_logdir()
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir2)

# cria um lista de callbacks

callbacks = [lr_scheduler,model_checkpoint_cb, tensorboard_cb,TqdmCallback(verbose=0)]

# Treina modelo
model_blsm.fit(X_train_min_max_blsm, y_train_blsm, epochs=180,batch_size=10,
                              validation_split=0.1,
                              verbose=0,
                              callbacks=callbacks)

# Carrega e Avalia o modelo
model_blsm = keras.models.load_model("models/churn_model_blsm_selu_nadam.h5") # retornar para o melhor modelo

# Avalia o modelo
print('\nAvaliação do modelo em Dados de teste')
churn_model_blsm_selu=model_blsm.evaluate(X_test_min_max_blsm, y_test)
churn_model_blsm_selu
```
