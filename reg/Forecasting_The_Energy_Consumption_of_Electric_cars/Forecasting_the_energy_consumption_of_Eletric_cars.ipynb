    Antes de iniciar este projeto, temos que lembrar quais métricas de desempenho devemos escolher para avaliar o modelo e assim devemos pensar cuidadosamente se ela é apropriada para o problema. Uma medida de típica de medir desempenho  preditivo de regressão é a **Raiz do Erro Quadrático Médio sigla (RMSE).** Ela dá uma ideia da quantidade de erros gerados pelo modelo em suas previsões (aos  valores previstos ŷi), com um peso maior para grandes erros.A figura mostra a fórmula da equação

> A métrica de desempenho mais importante da perspectiva da **Data Science** é a **Raiz do Erro Quadrático Médio ou RMSE. Dessa forma, esta é a métrica mais usada para comparar modelos de regressão**.

Além disso, outra medida é o **Erro Médio Absoluto (sigla MAE, em inglês)** também chamado de Desvio Médio Absoluto. É a soma da diferença absoluta entre previsões e valores reais .Dessa maneira, formenta a ideia de quão erradas estão nossas previsões. Portanto, valor igual a 0 indica que não há erro, sendo a previsão perfeita. A figura mostra a fórmula da equação 



<center>Image source: (GÉRON, 2021)</center>

- Sinônimo

    + **MAE**
    
**Portanto, tanto a RMSE quanto o MAE são formas de medir a distância entre dois <a href="https://lemas.furg.br/images/Apostilas/ebook_completo_novo_fabi.pdf">vetores:</a> o vetor das previsões e o vetor dos valores alvo. Várias medidas de distância, ou normas, são possíveis:**    

- Calcular a raiz de uma soma de quadrados (RMSE) corresponde à norma ℓ2, definida como norma  Euclidiana:


> **A distância Euclidiana corresponde à distância em linha reta entre dois pontos.**
    
    
- Calcular a soma de absolutos (MAE) corresponde à norma ℓ1, definida as vezes de norma  Manhattan


 >**A distância de Manhattan é a distância entre dois pontos em uma cidade.**

Coeficiente de determinação, também chamado de **estatística de R-quadrado ou R2**.Em geral, esse valor está entre 0 e 1 e representa o percentual da variância do alvo (target) com o qual os atributos contribuem, ou seja, é proporção de variância explicada pelo modelo. Valores maiores são melhores, mas, de modo geral, é difícil avaliar o modelo exclusivamente a partir dessa métrica. Um valor igual a 0,7 representa uma boa pontuação? Depende.Por outro lado, para um dado conjunto de dados, 0,5 pode ser uma boa pontuação, enquanto, para outro, um valor igual a 0,9 poderia ser ruim. Geralmente usamos esse número junto com outras métricas ou visualizações para avaliar um modelo.

É útil principalmente em usos explicativos da regressão nos quais se quer avaliar quão bem o modelo acomoda os dados. A fórmula para o coeficiente R2 é ilustra na figura a seguir.

###  1.3.1. Ideias chaves

As métricas mais importantes para avaliar um modelo são **(RMSE) e R-quadrado (R2)**.

O erro-padrão dos coeficientes pode ser usado para medir a confiabilidade da contribuição das variáveis para um modelo.

A regressão passo a passo é um modo de determinar automaticamente quais variáveis devem ser incluídas no modelo.**Em síntese, uma noção intuitiva do ajuste do modelo é: quão longe estão os valores estimados dos verdadeiros valores sobre os dados de treinamento? Em outras palavras, <font color="ff0000">quão grande é o erro a ser ajustado ao modelo? Certamente, queremos minimizar este erro quanto for possível</font>.**

De modo geral, o objetivo de um modelo de regressão é determinar a relação  entre uma variável aleatória 
$$  Y ∈ \mathbf{R}​​e​​um​​vetor​​x​​=​​(x1,. . ., xd)​​∈  \mathbf{R}^d $$ 

Mais especificmente, **busca-se estimar a função de regressão**

$$ r(x) := \mathbf{E}[ Y | X = x]$$

    A regressão é usada tanto para previsão quanto para explicação.De acordo, GORI (017) refina que um bom modelo de regressão é aquele capaz de explicar em grande medida o comportamento da variável dependente Y. Portanto, paraavaliar a qualidade de um ajuste, nada mais natural que medir a parcela do comportamento de Y explicada pela variável explanatória X, comparando-a com a associada aos resíduos do ajuste.


> <font color="#f0000">Em suma, os autores (PROVOST e FAWCETT , 2016) afirmam que existem muitas variedades de técnicas de ajuste de função, mas a maioria usa o mesmo modelo de estrutura linear: **uma simples soma ponderada dos valores de atributo**. Os parâmetros a serem ajustados pela mineração de dados são as ponderações dos atributos. Técnicas de modelagem linear incluem regressão linear tradicional, regressão logística e discriminantes lineares, como máquinas de vetores de suporte. Conceitualmente, a principal diferença entre estas técnicas é sua resposta a uma questão chave: O que exatamente queremos dizer com melhor ajuste dos dados? A qualidade do ajuste é descrita por uma “função objetiva”, e cada técnica utiliza uma função distinta. As técnicas resultantes podem ser bastante diversas</font>.

Logo, esses padrões são explicado **(EMMANUEL; MAUPONG; SEMONG; MPHAGO; TABONA, 2021)** como: 

- **Univariado:** O padrão de dados ausentes é univariado quando há apenas uma variável com dados ausentes. Esse padrão é raro na maioria das disciplinas e surge em estudos experimentais.
---

- **Monótono:** O padrão de dados ausentes é dito monótono se as variáveis nos dados puderem ser organizadas, esse padrão geralmente é associado a estudos longitudinais em que os membros desistem e nunca retornam . O padrão de dados monótono é mais fácil de lidar, pois os padrões entre os valores ausentes são facilmente observáveis.
---

- **Não monótono:** Este é um padrão de dados perdidos pelo qual a falta de uma variável não afeta a falta de quaisquer outras variáveis.


### 5.3.3. Tipos de dados ausentes

Existem alguns características de dados ausentes que você deve primeiro entender antes de abordá-los.

Nesma mesma linha de raciocínio, **(EMMANUEL; MAUPONG; SEMONG; MPHAGO; TABONA, 2021)** ressalta   principalmente, os tipos que causam valores perdidos nos dados influenciam algumas das suposições que sustentam a maioria das abordagens para lidar com dados perdidos e, assim, em literaturas os dados perdidos são definidos de acordo com esses tipos.Os autores refina a definição baseada nos estudos do autor **RUBIN**“[...] estabeleceram a teoria dos dados perdidos, categorizados por três mecanismos principais de falta, que são definidos dependendo dos dados disponíveis e faltantes.Para definir a falta, seja **Y**  uma matriz de todo o conjunto de dados que é decomposto em $$ \{Y_o\} e \{Y_m\}$$ que denotam os dados observados e ausentes. Seja R uma matriz de valor ausente definida por:”.


$$R := \left\{ \begin{array}{ll} 0, \quad \text{if} \; \text{Y}\; \text{é} \; \text{observado}\\ 1, \quad \text{if} \; \text{Y}\; \text{é} \; \text{ausente} \end{array}\right.$$

Seja **q** represente um vetor de valores que indicam a associação entre faltas em R e o conjunto de dados **Y** . Os mecanismos de valores ausentes são, portanto, definidos pela probabilidade de um valor ser observado ou ausente, conforme definido a seguir por **(EMMANUEL; MAUPONG; SEMONG; MPHAGO; TABONA, 2021)** tais como;


#### Missing completely at random (“Faltando completamente ao acaso”) MCAR

Isso ocorre quando as observações ausentes não dependem das medições observadas e não observadas. A probabilidade de MCAR é definida como:

$$p(R|q)$$

    Nesse cenário, os dados estão faltando apenas aleatoriamente e não há relação com outras variáveis no conj-unto de dados. Por exemplo, alguns dados podem estar faltando aleatoriamente devido a algum problema técnico ou devido a erro humano (KUMAR, 2021).

---
    
#### Missing at random  (“Faltando aleatoriamente”)  MAR


A probabilidade de um valor ausente no MAR está relacionada apenas aos dados observáveis. A probabilidade de MAR pode ser definida como:

$$p(R | Y_o,q )$$

    Nesse cenário, os dados ausentes têm alguma relação com outras variáveis no conjunto de dados. Por exemplo, em uma pesquisa, os campos de número de telefone podem não ser preenchidos pela maioria das mulheres devido a questões de segurança (KUMAR, 2021).

---


#### Missing not at random  (“Faltando não ao acaso”)  (MNAR)

Isso se refere a quando os dados ausentes não são MCAR nem MAR. Os dados perdidos dependem igualmente dos valores perdidos e observados. Nesse método, lidar com os valores ausentes geralmente é impossível, pois depende dos dados não vistos. A probabilidade MNAR é definida como:

$$p(R |Y_o, Y_m,q )$$	

A probabilidade de uma posição R estar ausente ou observada depende de $$ \{Y_o\} e  \{Y_m\}$$ Este tipo é aplicado principalmente em diferentes domínios predominantemente no domínio da (bio) medicina

    Nesse cenário, os dados não são perdidos aleatoriamente e a falta é atribuída aos dados que deveriam ser capturados. O MNAR é bastante complicado de identificar e lidar. Por exemplo, em um formulário de pesquisa, as pessoas ricas podem não preencher o campo Renda porque não gostariam de divulgá-lo (KUMAR, 2021).


Segundo autores destaca-se a conclusão referente aos autores **Graham e Dantan et al.**“[...] na maioria das vezes é impossível categorizar inequivocamente os dados ausentes nesses três mecanismos, pois imaginar que os dados ausentes não estão completamente relacionados a outras variáveis não ausentes é muito desafiador porque de uma forma ou de outra os **valores ausentes se relacionam a variáveis não ausentes**. No entanto, muitos pesquisadores relatam que a maneira mais fácil é completar todos os dados ausentes como MAR até certo ponto, já que MAR está no meio desse continuum”.


### 5.3.4. Abordagens de tratamento de valores ausentes


Assim como, nesta seção é o pouco desafiador aborda  todos os tipos de tratamento de valores ausentes diponsível em literaturas que você encontrará.Sendo assim, aboraremos algumas das mais simples e que pode ser aplicada em vários domínios, mas caso queira mergulha em águas profundas continuar lendo este artigo  <a href="https://journalofbigdata.springeropen.com/articles/10.1186/s40537-021-00516-9">A survey on missing data in machine learning</a>


Conforme mencionado anteriormente, existem várias técnicas e estratégias disponíveis para resolver o problema dos dados ausentes, no entanto, qual delas funciona melhor depende do seu conjunto de dados. **Não existe uma regra geral, então você terá que avaliar seu conjunto de dados e experimentar várias estratégias.**


####  Eliminação

Segundo (KUMAR, 2021) expôs nesta técnica a exclusão é considerada a abordagem mais simples, pois não há necessidade de tentar estimar o valor, a **linha ou coluna** que contém os dados ausentes é excluída completamente. Isso deve ser usado com cautela, pois você pode acabar perdendo informações importantes sobre os dados, pois introduz viés na análise principalmente quando os dados perdidos não são  distribuídos aleatoriamente. O conhecimento do domínio é bastante útil para decidir se descartar as colunas é a solução ideal para seu conjunto de dados. Por fim, O processo de exclusão pode ser realizado de duas maneiras, exclusão em pares ou em lista definidas por **(EMMANUEL; MAUPONG; SEMONG; MPHAGO; TABONA, 2021)**.


#### Exclusão de caso ou lista 


Na exclusão de lista, todos os casos que possuem um ou mais valores ausentes são removidos. A exclusão de lista tornou-se a escolha padrão ao analisar dados na maioria dos pacotes de software estatístico. No entanto, sob a suposição de que os dados não são MCAR, a lista resulta em viés. Embora, se as amostras de dados forem grandes o suficiente e a suposição MCAR for satisfeita, a exclusão em lista pode ser uma abordagem razoável. Se os dados amostrados não forem grandes ou a suposição MCAR não for satisfeita, a exclusão em lista não é a melhor abordagem a ser considerada. A exclusão de lista também pode resultar na perda de algumas informações importantes, especialmente quando os casos descartados são altos em números.

#### Exclusão em pares


Para mitigar a perda de informações ao fazer a exclusão em lista, pode-se usar a exclusão em pares. Isso ocorre porque a exclusão em pares é realizada de modo a reduzir as perdas que podem ocorrer na exclusão em lista. Isso é feito eliminando valores somente quando há um determinado ponto de dados necessário para testar se o valor assumido como ausente está de fato ausente. A fraqueza da exclusão de pares é que ela pode levar a uma matriz de <a href="https://www.lancaster.ac.uk/fss/courses/ling/corpus/Corpus3/3GLOSS.HTM#:~:text=intercorrelation%20matrix%3A%20This%20is%20calculated,distributions%20across%20the%20various%20samples">inter-correlação</a> que não é definida positiva, o que pode impedir análises adicionais, como o cálculo de estimativas de coeficientes. Em suma, a deleção em pares também é conhecida por produzir resultados de baixo viés para dados MCAR ou MAR .


#### Imputação 

O processo de imputação envolve a substituição de valores ausentes por alguns valores previstos. O conjunto de dados de valores não omissos é normalmente usado para prever os valores usados para substituir os valores omissos.Nesta técnica, os dados ausentes são preenchidos ou imputados por um substituto adequado e existem várias estratégias por trás disso, como as mais  sofisticadas (Imputação de regressão,Imputação hot-deck,Expectativa-maximização,Imputação múltipla).


A seguir, abordamos o métodos de imputação mais simples utilizados em literaturas.

#### Imputação Simples

A abordagem de imputação simples envolve a substituição de valores omissos para cada valor individual usando um atributo quantitativo ou atributo qualitativo de todos os valores não omissos. Com a imputação simples, os dados ausentes são tratados por diferentes métodos, como **moda, média ou mediana dos valores disponíveis**. Na maioria dos estudos são utilizados métodos de imputação simples devido à sua simplicidade e que podem ser usados como uma técnica de referência fácil. No entanto, métodos de imputação simples podem produzir viés ou resultados irreais em conjuntos de dados de alta dimensão. Além disso, com a geração de big data emergindo, esse método parece ter um desempenho ruim e, portanto, é inadequado para ser implementado em tais conjuntos de dados.

Todas as ocorrências de valores ausentes serão preenchidas com a estratégia de imputação selecionada.

- Se **“mean”**, substitui os valores ausentes usando a média ao longo de cada coluna.Observa-se que só pode ser usado com dados numéricos.No entanto, temos que ser cautelosos aqui porque se os dados na coluna contiverem valores discrepantes, sua média será enganosa

- Se **“median”**, substitui os valores ausentes usando a mediana em cada coluna. Novamente, só pode ser usado com dados numéricos.

- Se **“most_frequent”**, substitui os valores ausentes usando o valor mais frequente em cada coluna e pode ser usado em colunas categóricas ou  numéricos.

- Se **“constant”**, nessa abordagem, os dados ausentes são substituídos por um valor constante. Isso pode ser usado com colunas numéricas e categóricas

Em suma, tem os métodos de imputação inspirados em **Machine Learning** já entramos em outro universo, que são técnicas sofisticadas que envolvem principalmente o desenvolvimento de uma abordagem preditiva para lidar com valores ausentes usando aprendizado não supervisionado ou supervisionado.

```python
def plot_missing_values(df):
    """Para cada coluna com valores ausentes, plote a proporção que está faltando."""
    
    data = [(col, df[col].isna().sum() / len(df)*100)
            for col in df.columns if df[col].isnull().sum() > 0]
    
    col_names = ['column', 'percent_missing',]
    
    missing_df = pd.DataFrame(data, columns=col_names).sort_values('percent_missing')
    
    # Cria a figura com orientação horizontal
    fig = go.Figure(go.Bar(x= missing_df['percent_missing'], y=missing_df['column'], orientation = "h")) 

    # Layout
    fig.update_layout(title_text = 'Porcentagem de valores ausentes nas colunas',
                  title_x = 0.5,
                  #font_color ="#000000",
                  xaxis_title = "%", 
                  yaxis_title = "Variável(is) ",
                  template ="plotly_white",
                  font=dict(color="black"))
    
    # ajuste dos espaço em branco
    fig.update_yaxes(ticksuffix = "   ")
    fig.show()
```

#### Definindo os pipelines

Assim sendo, definimos dois pipelines de pré-processamento de recursos; um para variáveis numéricas (num_pipeline) e outro para variáveis categóricas (cat_pipeline). 

- **num_pipeline** importa o método da classe SimpleImputer para imputação de dados ausentes e MinMaxScaler() para dimensionamento de dados. 

- **cat_pipeline** importa o método da classe em **SimpleImputer** para imputação de dados ausentes e OneHotEncoder para codificação de dados categóricos como dados numéricos.

- Por fim, combinar  esses dois pipelines usando **ColumnTransformer** para aplicá-los a um conjunto diferentes  colunas.

```python
#selecione os atribuitos númericos

#observe-que vamos seleciona a variável target (Energy_consumption) pois ela contém valores NAN
num_attribs = df.select_dtypes(include="number").columns.tolist()

# selecione os atributos categorical
cat_attribs = df.select_dtypes('object').columns.tolist()
```

```python
# construindo um pipeline para pré-processar os atributos numéricos:
num_pipeline = Pipeline([
    ('num_imputer', SimpleImputer(strategy='median',add_indicator=False)),
    ('min_max', MinMaxScaler()),
])


# Criando um  pipeline  para Transformação One Hot Encoding
cat_pipeline = Pipeline([
    ('cat_imputer',SimpleImputer(strategy='most_frequent',fill_value='Missing')),
    ('ohe',OneHotEncoder(handle_unknown='ignore',sparse=False))
],
)
```

Definir remainder='passthrough' significará que todas as colunas não especificadas na lista de “ transformers ” serão passadas sem transformação, em vez de serem descartadas.

a função a seguir é pouco complicada, basicamente ela retorna o nomes das variáveis dos dados transformados ao final da aplicação.O código a seguir é sobre a <a href="https://github.com/scikit-learn/scikit-learn/issues/12525">Cannot get feature names after ColumnTransformer #12525</a>

```python
from skimpy import clean_columns
from sklearn.utils.validation import check_is_fitted

def get_column_names_from_ColumnTransformer(column_transformer, clean_column_names=True, verbose=True):  

    """
    Reference: Kyle Gilde: https://github.com/kylegilde/Kaggle-Notebooks/blob/master/Extracting-and-Plotting-Scikit-Feature-Names-and-Importances/feature_importance.py
    Description: Get the column names from the a ColumnTransformer containing transformers & pipelines
    Parameters
    ----------
    verbose: Bool indicating whether to print summaries. Default set to True.
    Returns
    -------
    a list of the correct feature names
    Note:
    If the ColumnTransformer contains Pipelines and if one of the transformers in the Pipeline is adding completely new columns,
    it must come last in the pipeline. For example, OneHotEncoder, MissingIndicator & SimpleImputer(add_indicator=True) add columns
    to the dataset that didn't exist before, so there should come last in the Pipeline.
    Inspiration: https://github.com/scikit-learn/scikit-learn/issues/12525
    """

    assert isinstance(column_transformer, ColumnTransformer), "Input isn't a ColumnTransformer"
    
    check_is_fitted(column_transformer)

    new_feature_names, transformer_list = [], []

    for i, transformer_item in enumerate(column_transformer.transformers_): 
        transformer_name, transformer, orig_feature_names = transformer_item
        orig_feature_names = list(orig_feature_names)

        if len(orig_feature_names) == 0:
            continue

        if verbose: 
            print(f"\n\n{i}.Transformer/Pipeline: {transformer_name} {transformer.__class__.__name__}\n")
            print(f"\tn_orig_feature_names:{len(orig_feature_names)}")

        if transformer == 'drop':
            continue

        if isinstance(transformer, Pipeline):
            # if pipeline, get the last transformer in the Pipeline
            transformer = transformer.steps[-1][1]

        if hasattr(transformer, 'get_feature_names_out'):
            if 'input_features' in transformer.get_feature_names_out.__code__.co_varnames:
                names = list(transformer.get_feature_names_out(orig_feature_names))
            else:
                names = list(transformer.get_feature_names_out())
        elif hasattr(transformer, 'get_feature_names'):
            if 'input_features' in transformer.get_feature_names.__code__.co_varnames:
                names = list(transformer.get_feature_names(orig_feature_names))
            else:
                names = list(transformer.get_feature_names())

        elif hasattr(transformer,'indicator_') and transformer.add_indicator:
            # is this transformer one of the imputers & did it call the MissingIndicator?

            missing_indicator_indices = transformer.indicator_.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]
            names = orig_feature_names + missing_indicators

        elif hasattr(transformer,'features_'):
            # is this a MissingIndicator class? 
            missing_indicator_indices = transformer.features_
            missing_indicators = [orig_feature_names[idx] + '_missing_flag'\
                                  for idx in missing_indicator_indices]

        else:

            names = orig_feature_names

        if verbose: 
            print(f"\tn_new_features:{len(names)}")
            print(f"\tnew_features: {names}\n")

        new_feature_names.extend(names)
        transformer_list.extend([transformer_name] * len(names))

    transformer_list, column_transformer_features = transformer_list, new_feature_names

    if clean_column_names:
        new_feature_names = list(clean_columns(pd.DataFrame(columns=new_feature_names)).columns)
    
    return new_feature_names
```

**<a href="https://numpy.org/doc/stable/reference/generated/numpy.argpartition.html" target="blank">numpy.argpartition()**</a> 
 
---
A função é usada para criar uma cópia particionada indireta do array de entrada com seus elementos reorganizados de tal forma que o valor do elemento na k-ésima posição esteja na posição que estaria em um array ordenado. Todos os elementos menores que o k-ésimo elemento são movidos antes desse elemento e todos iguais ou maiores são movidos para trás dele. A ordenação dos elementos nas duas partições é indefinida. Ela retorna uma matriz de índices da mesma forma que arr, ou seja, arr[index_array] gera uma partição de arr.

```python
# função par manter lista de idx que devem ser mantidos em ordem ordenado, é usada para realizar uma partição indireta.
def indices_of_top_k(arr, k):
    return np.sort(np.argpartition(np.array(arr), -k)[-k:])

class TopFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, feature_importances, k):
        self.feature_importances = feature_importances
        self.k = k
        
    def fit(self, X, y=None):
        self.feature_indices_ = indices_of_top_k(self.feature_importances, self.k)
        return self
    
    def transform(self, X):
        return X[:, self.feature_indices_]
```

### <font color="red"> 7.3.1. Correlação vs Colinearidade vs Multicolinearidade</font>

Vamos simplificar as diferenças de  correlação,colienaridade e multicolinearidade, pois este é o cerne do conceito fundamental para distingui-lás


| |Correlação |Colinearidade|Multicolineariedade|

|**Definição**|**Correlação refere-se à relação linear entre 2 variáveis**|**A colinearidade refere-se a um problema ao executar um modelo de regressão em que 2 ou mais variáveis independentes (também conhecidas como preditores) têm uma forte relação linear**| **A multicolinearidade é um caso especial de colinearidade onde existe uma forte relação linear entre 3 ou mais variáveis independentes, mesmo que nenhum par de variáveis tenha uma alta correlaçã**|

**<font color="red"> 7.3.2. Termos-chave</font>**

**A forte correlação entre 2 variáveis independentes causará um problema na interpretação do modelo linear e esse problema é chamado de colinearidade.**

De fato,colinearidade é um termo mais geral que também abrange casos em que 2 ou mais variáveis independentes estão linearmente relacionadas entre si. Portanto, a colinearidade pode existir porque um par de preditores está correlacionado ou porque 3 ou mais preditores estão linearmente relacionados entre si. **Este último caso é por vezes referido como multicolinearidade, ou seja,a multicolinearidade pode causar também impactos na interpretação da importância dos atributos e dos coeficientes em alguns modelos**.


De acordo (GORI, 2017) explica o conceito original  por meio das palavras **Ragnar Frisch**. De acordo com ele, *O termo multicolinearidade foi originalmente proposto por Ragnar Frisch, em 1934, para designar a relação colinear exata entre duas ou mais variáveis independentes*. Atualmente, refere-se a um conceito mais amplo, de inter-relação entre as variáveis independentes, mas não de maneira exata.

Observa-se que (GORI, 2017) refina que a multicolinearidade apenas dificulta a identificação dos efeitos isolados na amostra e, caso esta não seja suficientemente representativa dos inúmeros comportamentos das variáveis (represente variações conjuntas e isoladas dos regressores), não será suficiente para estimar coeficientes significativos para os efeitos marginais. Assim, o impacto da multicolinearidade sobre as estimativas dos coeficientes dependerá também de outros fatores, **como o tamanho da amostra e a variabilidade do regressor**. Por esse motivo, muitos autores argumentam que a multicolinearidade trata-se, na verdade, de um problema de **micronumerosidade**, ou seja, que a amostra não seria grande o suficiente para representar significativamente o relacionamento entre as variáveis em questão

>*O conceito da micronumerosidade foi sugerido por Arthur Goldberger (1991), argumentando que o problema da multicolinearidade deve-se, na verdade, à pequena variabilidade dos regressores observados em uma amostra de tamanho insuficiente. Segundo o autor, amostras de tamanho pequeno e baixa variabilidade dos regressores causam problemas tão graves quanto o da multicolinearidade.(GORI, 2017)*.

**Estatísticas conflitantes:** um **R²** elevado em um modelo com poucas **estatísticas t significativas**. Em um caso extremo, teríamos uma **estatística F** significativa sem qualquer estatística individualmente **t** significativa. Enquanto o **R²** e a estatística F analisam o comportamento conjunto das variáveis independentes, as estatísticas t analisam o efeito isolado de cada uma. Assim, poderíamos ter uma contribuição conjunta significativa sem que as contribuições marginais.


**Transformar as variáveis:** a multicolinearidade pode ser eliminada transformando-se as variáveis independentes. Por exemplo, se estimarmos o preço de venda da soja com base na área e quantidade produzida, teremos provavelmente uma relação de colinearidade entre área e quantidade produzida. Entretanto, se substituirmos ambas as variáveis independentes pela variável produtividade (produção/área), estaríamos eliminando esse problema.


**Exclusão de variáveis:** uma solução simples, mas perigosa, é a
exclusão de uma ou mais variáveis que apresentam multicolinearidade. A exclusão de variáveis essenciais para
compreensão do problema pode, entretanto, gerar o chamado **viés
de especificação**, que é a falha na formulação apurada das relações entre a variável dependente e independente


**Inclusão de variáveis:** a inclusão de regressores que não estejam relacionados às demais variáveis independentes, mas sim a variável dependente, reduzirá a variância da regressão (σ2
), sem aumentar o FIV. Em amostras pequenas, entretanto, a redução dos graus de liberdade dos resíduos pode comprometer a significância das estimativas.

>**Exclusão de variáveis:** exclusão de variáveis independentes, usualmente coloca o pesquisador em um dilema entre a variância e o viés dos estimadores. Suponha, por exemplo, que na real relação da população Y dependa de X1 e X2.

### 7.5. Variance Inflation Factor - VIF


Assim de fato, a correlação seja uma medida útil do grau de associação entre duas variáveis, não explica algumas questões fundamentais, como a colinearidade também pode ocorrer entre 3 variáveis ou mais, MESMO **quando nenhum par de variáveis é altamente correlacionado (uma situação muitas vezes referida como “multicolinearidade”)**, a matriz de correlação não pode ser usada para detectar todos os casos de colinearidade.Entretano, calculando o **Variance Inflation Factor** removemos as correlações perfeitas que não nos permitiam inverter adequadamente  uma matriz e obter os coeficientes. 

No entanto, estamos omitindo o fato de que algumas das correlações ainda podem ser  grandes e podemos removê-las. Existem várias maneiras de fazer isso: uma opção é  usar os chamados *fatores de inflação de variância* (VIFs); esses valores são proporções que mostram quanto maiores os erros padrão ficam entre  um modelo que contém apenas uma variável e um modelo que contém todas elas. 

O VIF é calculado pela seguinte expressão:


$$ VIF_1 =  \frac {1} {1-R^2}$$ 

Na equação , R² é o coeficiente de determinação do modelo de regressão linear, temos:

$$ X_1 ​​como​​variável​​dependente $$ 
$$ X_2 , X_3 , X_4 , X_n,​​... ​​como​​variáveis​​independentes$$ 


Porque R² é um número entre 0 e 1:

+ Quando R²  está próximo de 1 (X 2 , X 3 , X 4 , … são altamente preditivos de X 1 ): o VIF será muito grande

+ Quando R²  está próximo de 0 (X 2 , X 3 , X 4 , … não estão relacionados a X 1 ): o VIF estará próximo de 1

    Se um VIF for grande, significa que não obteremos uma estimativa precisa do efeito  e, se tivermos mais dados, a estimativa mudará bastante. 


Nesse sentido,VIF  é como um indicador que significa,tendo como regra geral, um VIF > 10 é um sinal de multicolinearidade. No entanto, isso é um pouco uma simplificação excessiva. Também devemos prestar atenção aos  valores de VIF entre 5 e 10. 


```python
# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]

# view
print (vif_data)
```

### 9.1. Selecionando e Treinando Modelo

#### Avaliando melhor com a Cross Validation

Uma boa alternativa é utilizar o método **k-fold** (“seções/dobras”) de **Cross Validation** (“Validação Cruzada”) do Scikit-Learn. O código a seguir executa a *Validação Cruzada K-fold*:  divide aleatoriamente o conjunto de treinamento em 10 subconjuntos distintos chamados de partes (folds), então treina e avalia o modelo  10 vezes escolhendo uma parte (fold) diferente a cada uma delas para avaliação e treinando nas outras nove partes. O resultado é um array contendo as 10 classificações de avaliação.

```python
# Definindo os valores para o número de folds

num_folds = 10

# Criando uma lista, para armazenar os modelos que serão utilizados.
models = []
models.append(('ET3R', ExtraTreesRegressor()))
models.append(('RFR',  RandomForestRegressor()))
models.append(('GBTR', GradientBoostingRegressor()))
models.append(('XGBR', xgb.XGBRegressor()))
models.append(('XGBRFR', xgb.XGBRFRegressor()))


# Criando Listas, para armazenar os resultados e os nomes, de cada um dos algoritmos testados.

results = []
names   = []

# Criando um Dataframe, para armazenar a média e o desvio-padrão, de cada um dos algoritmos testados.

means   = pd.DataFrame(columns = ['mean', 'std'])

for name, model in models:
        
        # Instanciando um objeto da Kfold para criar os folds.

        kfold = KFold(n_splits = num_folds)
        
        # Treinando o modelo com Cross Validation.

        cv_results = np.sqrt(-cross_val_score(model, X_train_min_max, y_train, cv = kfold,scoring="neg_mean_squared_error"))
        
        # Adicionando os resultados gerados a lista de resultados.

        results.append(cv_results)
        
        # Adicionando o nome do modelo avaliado a lista de nomes.

        names.append(name)

        # Adicionando a média e o desvio-padrão dos resultados gerados, pelo modelo analisado ao Dataframe de médias.

        means = means.append(
            pd.DataFrame (
                data    = [[cv_results.mean(), cv_results.std()]], 
                columns = ['mean', 'std'], 
                index   = [name]
            )
        )           
```

**Os recursos de validação cruzada do Scikit-Learn esperam uma função de utilidade (quanto maior, melhor) do que uma função de custo (quanto menor, melhor); de modo, a função score é de fato oposto do MSE (ou seja, um valor negativo), é por essa razão  no qual o código anterior calcula (-scores), antes de calcular a raiz quadrada.**




<font color="#f000">**A raiz do erro quadrático médio negativo** ('neg_mean_squared_error' no **GridSearch ou  Cross-Validation**) também avalia o erro do modelo em termos do alvo. No entanto, como ele calcula a média do quadrado dos erros antes de calcular a raiz quadrada, erros maiores são penalizados. Se quiser penalizar erros maiores, essa é uma boa métrica a ser usada: por exemplo, se estar errado em oito for mais do que duas vezes pior do que estar errado por quatro.Além disso,  é uma das métricas de desempenho mais importante da perspectiva da ciência de dados é a raiz quadrada do erro quadrático médio, ou RMSE, que é a raiz quadrada do erro quadrado médio nos valores previstos ŷi</font>

Como no caso do erro médio absoluto (MAE), essa medida não é capaz de informar quão ruim é um modelo, mas pode ser usada para comparar dois modelos. Caso você suponha que os erros estão distribuídos de modo normal, essa será uma boa opção.

### 10.1.2. Hiperparameters


> O que significa ajuste de hiperparâmetro?

Segundo ((TOUSHIK) defende que no aprendizado de máquina, **a otimização ou ajuste de hiperparâmetros é o problema de escolher um conjunto de hiperparâmetros ideais para um algoritmo de aprendizado.** Um hiperparâmetro é um parâmetro cujo valor é usado para controlar o processo de aprendizado, ou seja, é uma configuração externa ao modelo e cujo valor não pode ser estimado a partir de dados. Por outro lado, os valores de outros parâmetros (normalmente os pesos dos nós) são aprendidos. O processo de ajuste de hiperparâmetro é uma caminhada na corda bamba para alcançar um equilíbrio entre **underfitting e overfitting**. Underfitting é quando o modelo de aprendizado de máquina não consegue reduzir o erro para o conjunto de teste ou treinamento.

--- 
> Suas principais características são:

- Eles são frequentemente usados em processos para ajudar a estimar os parâmetros do modelo.

- Eles são frequentemente especificados pelo pesquisador. 

- Eles geralmente podem ser configurados usando heurísticas.

- Eles são frequentemente ajustados para um determinado problema de modelagem preditiva.

---

> O que é exemplo de ajuste de hiperparâmetro?

Observa-se que (TOUSHIK) ressaltar alguns exemplos de hiperparâmetros de modelo incluem: A penalidade na regularização do Classificador de Regressão Logística, ou seja, L1 ou L2. A taxa de aprendizado para treinar uma rede neural. Os hiperpârmetros C e sigma para máquinas vetoriais de suporte.


> Por que usamos o ajuste de hiperparâmetros?

Note-se,(TOUSHIK) refina a importância do ajuste de hiperparâmetros é uma parte essencial do controle do comportamento de um modelo de aprendizado de máquina. Se não ajustarmos corretamente nossos hiperparâmetros, nossos parâmetros de modelo estimados produzirão resultados abaixo do ideal, pois não minimizarão a função de perda. **Isso significa que nosso modelo comete mais erros.**

--- 
Dessa forma, não temos como conhecer o melhor valor para um hiperparâmetro em um determinado problema. Podemos usar regras básicas, copiar valores usados em outros problemas ou  procurar o melhor valor por tentativa e erro. Quando um algoritmo de aprendizagem de máquina é ajustado para um problema  específico, usando por exemplo **grid search ou random search**, você está ajustando os hiperparâmetros do modelo para descobrir os parâmetros do modelo que resultam nas previsões mais precisas.Assim acresce-se que, quando se trata de otimizar um modelo, você testa diversas variações de hiperparâmetros nesse modelo com todas as combinações possíveis de hiperparâmetros em um espaço de busca.


Em poucas palavras, hiperpârametros se refere a números que você usa para acertar ou ajustar o desempenho de um modelo estatístico ou de aprendizado de máquina.

### <center>Booster parameters</center>

Então esta parte descata-se a qual reforço você escolhará. Temos a seguintes propriedades sobre boosters baseados em árvores destacam-se:

#### 1. eta [default=0.3]

- Semelhante ao learning_rate; trabalha com generalização.
- Valores finais típicos a serem usados: 0,01-0,2

Segundo (ANDREW e BRUCE, 2019) o fator de encolhimento ou taxa de aprendizagem está entre 0 e 1 aplicado a α no algoritmo de boosting. O padrão é 0,3, **mas, para dados ruidosos, recomenda-se valores menores (por exemplo, 0,1)**. Nesse mesmo sentido, (HARRIOSON ,2020) apura que após cada passo de boosting, (melhoria) os pesos recém-adicionados são escalados de acordo com esse fator. Quanto menor o valor, mais conservador será, mas também serão necessárias mais árvores para convergir. Na chamada ao dados de train, você pode passar um parâmetro learning_rates, que é uma lista de taxas em cada rodada (isto é, [.1] * 100 + [.05] * 100)


#### 2. min_child_weight [default=1]

- Define **a soma mínima dos pesos de todas as observações necessárias em um filho**. É usado para **controlar o ajuste excessivo (overfitting)**. Valores mais altos impedem que um modelo aprenda  relações que podem ser altamente específicas para a amostra específica selecionada para uma árvore.


- Valores muito altos podem levar a um ajuste insuficiente, portanto, ele deve ser ajustado usando CV.


#### 3. max_depth [padrão=6]

- A **profundidade máxima** de uma árvore também é usada para **controlar o ajuste excessivo**, pois uma profundidade maior permitirá que o modelo aprenda relações muito específicas para uma amostra específica. Ou seja, ao contrário da floresta aleatória **(random forest)**, que ajusta árvores muito profundas, o **boosting costuma ajustar árvores rasas**. Isso tem a vantagem de evitar interações falsas e complexas no modelo, as quais podem ser resultado de dados ruidosos (ANDREW e BRUCE, 2019).


- Deve ser ajustado usando CV. Valores típicos: 3-10


#### 4. max_leaf_nodes

- O número máximo de nós terminais ou folhas em uma árvore.

- Como as árvores binárias são criadas, uma profundidade de 'n' produziria um máximo de 2^n folhas, ou seja, em que  cada nó da árvore tem  no máximo 2 filhos. Se isso for definido, o GBM ignorará max_depth.


#### 5. gamma [default=0]

- Um nó é dividido apenas quando a divisão resultante fornece uma redução positiva na função de perda.**Gamma especifica a redução de perda mínima necessária para que ocorra uma divisão.**

- Torna o algoritmo conservador. Os valores podem variar dependendo da função de perda e devem ser ajustados. Pode ser usado para controlar o overfitting.

Observa-se (HARRISON, 2020) refina que gama controla a poda (pruning). Assim como, pode varia de 0 a infinito, a redução de perda mínima necessária para separar mais uma folha. Quanto maior o valor de gama, mais conservador será. Se as pontuações de treinamento e de teste divergirem, insira um número maior (em torno de 10). Se scores de treinamento e teste estiverem próximas, utilize um número menor.

#### 6. max_delta_step [default=0]

- No passo **delta máximo, permitimos que a estimativa de peso de cada árvore seja**. Se o valor for definido como 0, significa que não há restrição. Se estiver definido com um valor positivo, pode ajudar a tornar a etapa de atualização mais conservadora.

- Isso geralmente não é usado, mas sim, usado para classes desbalanceadas, pois deixa as atualizações mais conservadoras com valores de 1 a 10.


#### 7. subsample [default=1]

- Denota a fração de observações a serem amostras aleatórias para cada árvore.

- Valores mais baixos tornam o **algoritmo mais conservador e evitam overfitting**, mas valores muito pequenos podem levar a underfitting.

- Valores típicos: 0. 5 - 1

- Conforme discutido, a subsample especifica a fração de instâncias de treinamento a serem usadas para treinar cada árvore. Por exemplo, se subsample=0.75, cada árvore é treinada em **75% das instâncias de treinamento selecionadas aleatoriamente.**Como você provavelmente já deve ter adivinhado, essa técnica troca um **viés maior por uma variância menor.** Ela também acelera substancialmente o treinamento. Essa técnica se chama **gradiente estocástico boosting.**



#### 8. colsample_bytree [padrão=1]

- Mostra a fração de colunas a serem amostras aleatórias para cada árvore. Colsample_bytree menor fornece regulação adicional.

- Valores típicos: 0.5 -1

#### 9. colsample_bylevel [default=1]

- Indica a proporção de **subamostra de colunas para cada divisão, em cada nível**.

#### 10. colsample_bynode [default=1]

- Indica a proporção de **subamostra de colunas para cada divisão, em cada nó**.


#### 11. alpha [default=0]


- **Termo  de regularização L1.**

- A regularização L1 força os pesos dos recursos não informativos a serem zero, subtraindo uma pequena quantidade do peso em cada iteração e, assim, tornando o peso zero, eventualmente. Também é chamado de regularização por esparsidade <a href="https://developers-google-com.translate.goog/machine-learning/crash-course/regularization-for-sparsity/l1-regularization?_x_tr_sl=en&_x_tr_tl=pt&_x_tr_hl=pt-BR&_x_tr_pto=sc">regularização por esparsidade</a>. - Além do mais, é mais suave que alfa. Assim sendo, aplica-se aos pesos das folhas.

#### 12. scale_pos_weight [default=1]

```python
def xgb_model_helper(PARAMETERS,V_PARAM_NAME=False,V_PARAM_VALUES=False,BR=10):
    
    # Cria uma matrix temporária em formato de bit do conjunto de dados a ser treinados
    temp_dmatrix = xgb.DMatrix(data =X_train, label = y_train)
    
    #Check os parâmetros a ser utilizados
    if V_PARAM_VALUES==False:
        cv_results = xgb.cv(dtrain=temp_dmatrix, nfold=5,num_boost_round=BR,params=PARAMETERS, as_pandas=True,seed=123)
        return cv_results
    
    else:
        # Criando uma Lista, para armazenar os resultados e os nomes, de cada uma das métricas.
        results = []
        
        # Percorre a lista de parâmetros
        for v_param_value in V_PARAM_VALUES:
            
            # Adicionando o nome dos parâmetros avaliado a lista de nomes.
            PARAMETERS[V_PARAM_NAME] = v_param_value
            
            # Treinando o modelo com Cross Validation.
            cv_results = xgb.cv(dtrain= temp_dmatrix, nfold=5, num_boost_round=BR, params=PARAMETERS, as_pandas= True, seed =123)
           
            # Adicionando os resultados gerados a lista de resultados.
            results.append((cv_results["train-mae-mean"].tail().values[-1], cv_results["test-mae-mean"].tail().values[-1]))  # .tail().values[-1] captura somente as colunas
            
        # zip “pareia” os elementos de uma série de listas, tuplas ou outras sequências para criar uma lista de tuplas:

        # Adicionando a média da AUC e o desvio-padrão dos resultados gerados, pelo modelo analisado ao Dataframe de médias.
        data = list(zip(V_PARAM_VALUES, results))
        print(pd.DataFrame(data,columns=[V_PARAM_NAME,"mae"]))
        
        return cv_results
```

### 10.8.1. L1 vs L2 Regularização

De acordo (DEVELOPERS-GOOGLE) esclarece alguns fundamentos específicos sobre as regulariação L2 e L1 que apresentadas da seguintes formas:


L2 e L1 penalizam os pesos de forma diferente:

- L2 penaliza weight².
- L1 penaliza |weight|.

Consequentemente, L2 e L1 têm derivadas diferentes:

- A derivada de L2 é **2 * weight**

A derivada de L1 é **k (uma constante, cujo valor é independente do peso).**

Nesssa mesma linha de trabalho, (DEVELOPERS-GOOGLE) defende que você pode pensar na **derivada de L2 como uma força que remove x% do peso todas as vezes**. DEVELOPERS-GOOGLE ressaltar um conceito importante por meio das palavras do autor **ZENO** de acordo com ele *mesmo se você remover x por cento de um número bilhões de vezes, o número diminuído ainda nunca chegará a zero.* (Zeno estava menos familiarizado com as limitações de precisão de ponto flutuante, que poderiam produzir exatamente zero.) De qualquer forma, **L2 normalmente não leva os pesos a zero.**

Conseqüentemente, você pode pensar na **derivada de L1 como uma força que subtrai alguma constante do peso todas as vezes**. No entanto, graças aos **valores absolutos, L1 tem uma descontinuidade em 0, o que faz com que os resultados de subtração que cruzam 0 sejam zerados**. Por exemplo, se a subtração forçaria um peso de +0,1 a -0,2, L 1 definirá o peso exatamente 0. <a href ="https://pt.wikipedia.org/wiki/Eureka_(exclama%C3%A7%C3%A3o)">Eureka</a>, L1 zerou o peso.

>A regularização L 1  penalizando o valor absoluto de todos os pesos, acaba sendo bastante eficiente para modelos largos.

###  R²

<font color="#f000"> Um bom modelo de regressão é aquele que ajuda a explicar em grande medida a variablidadede Y. O coeficiente de determinação (R²) é uma métrica de regressão simples e muito útil de qualidade do ajuste, onde estima a proporção da variabilidade da variável dependente que é explicada pela variável independente do modelo de regressão.O **R²** assumirá  valores  entre  0 e 1, quando a variabilidade explicada pelo modelo for nula, e um, quando 100% da variabilidade total for explicada pelo modelo que representa o percentual da variância do alvo (target) com o qual os atributos contribuem. Assim como, valores maiores são melhores, mas, de modo geral, é difícil avaliar o modelo exclusivamente a partir dessa métrica.Bem como, um valor igual a 0.7 representa uma boa pontuação? Depende.Porém, para um dado conjunto de dados, 0.5 pode ser uma boa pontuação, enquanto, para outro, um valor igual a 0.9 poderia ser ruim. Geralmente usamos esse número junto com outras métricas ou visualizações para avaliar um modelo.Em outras palavras conforme ilustra a figura a seguir simplifica o conceito da equação do coeficiente determinação</font>


<center>Source Image: (GORI, 2017)</center>

O coeficiente de correlação, o **R²** pode ter um valor negativo quando o modelo selecionado não segue a tendência dos dados, levando a um ajuste pior do que a linha horizontal. Geralmente é o caso quando há restrições no intercepto ou na inclinação da linha de regressão linear.Observe que **R²** nem sempre é o quadrado de qualquer coisa, portanto, pode ter um valor negativo sem violar nenhuma regra matemática. 

Nesse mesmo sentido, conforme defende (GORI, 2017) o coeficiente de correlação, o **R²** é uma medida de relação linear.Assim, embora variáveis sem relação apresentem R² nulo ou próximo
de 0, a recíproca não é necessariamente verdadeira, já que a relação entre **Y e X** pode ser não linear.Outra precaução afirma (GORI, 2017) <mark>em relação à análise do **R²** refere-se ao fato de valores baixos para esta estatística não necessariamente significarem um ajuste insatisfatório. Algumas variáveis, como, por exemplo, a riqueza de uma pessoa, são muito difíceis de serem determinadas quantitativamente. Mesmo uma baixa contribuição de um fator explanatório pode nos dar uma importante fonte de informação.</mark>

#### R2 mais alto é melhor?

A interpretação mais comum do **R²**  é quão bem o modelo de regressão se ajusta aos dados observados, ou seja, mede quão perto os dados estão da linha de regressão. Por exemplo, um **R²**  de 70% mostra que 70% dos dados se ajustam ao modelo de regressão. Geralmente, um **R²** mais alto indica um melhor ajuste para o modelo 

### O que é mean square error negativo MSE?

Em estatística, o erro quadrado médio (**MSE**) não pode retornar valores negativos. Embora a diferença entre um valor e a média possa ser negativa, esse valor negativo é elevado ao quadrado. Portanto, todos os resultados são positivos ou zero.


    Lembre-se, quando elevamos um número positivo ou um número negativo ao quadrado, temos como resultado um número positivo.Vejamos os exemplo a seguir


- $$ (5)^2 = (5) * (5) = 25 $$



- $$ (\frac{1}{3} )^2 = (\frac{1}{3} )^2 * (\frac{1}{3})^2 = (\frac{1}{9}) $$  



- $$ (-7)^2 = (-7) * (-7) = 49 $$


- $$ (- \frac {2}{3} )^2 = (- \frac{2}{3} )^2 * (- \frac  {2}{3})^2 = (\frac{4}{9}) $$  



###  O  mean square error  pode ter valores negativos?

O **MSE** é uma função de risco (perda), correspondente ao valor esperado da perda de erro ao quadrado. O fato de o **MSE** ser quase sempre estritamente positivo (e não zero) é devido à aleatoriedade ou porque o estimador não leva em conta informações que poderiam produzir uma estimativa mais precisa.


###  O que é um bom mean square error ?

**Não há valor correto para MSE**. Simplificando, quanto menor o valor, melhor e 0 significa que o modelo é perfeito 100% significa correlação perfeita. No entanto, existem modelos com baixo **R²** que ainda são bons modelos.

###  O que significa precisão negativa?

E seu valor negativo denota que a linha de regressão não está seguindo a tendência dos dados, ou seja, **significa que o modelo não é bom**



###  O que é erro absoluto negativo (MAE)?

Como o próprio nome indica, MAE negativo é simplesmente o negativo do MAE , que (MAE) é por definição uma quantidade positiva. E como MAE é uma métrica de erro, ou seja, quanto menor melhor, MAE negativo é o oposto: um valor de -2,6 é melhor que um valor de -3,0 .


###  Um RMSE maior ou menor é melhor?

O RMSE é a raiz quadrada da variância dos resíduos.Valores 
mais baixos de RMSE indicam **melhor ajuste**. O **RMSE** é uma boa medida da precisão com que o modelo prevê a resposta e é o critério mais importante para o ajuste se o objetivo principal do modelo for a previsão.Em outras palavras, é métrica mais importante para avaliar e comparar modelos de regressão, bem como  R²


### RMSE é melhor que o MSE?

De fato, quanto menor o mean squared error **(MSE)**, mais próximo o ajuste está dos dados. O MSE tem as unidades ao quadrado de tudo o que é plotado no eixo vertical. O **RMSE** é diretamente interpretável em termos de unidades de medida e, portanto, é uma medida melhor de qualidade de ajuste do que um coeficiente de correlação .


###   O que o RMSE mostra?

Root Mean Square Error **(RMSE)** é o desvio padrão dos resíduos (erros de previsão). Resíduos são uma medida de quão longe os pontos de dados da linha de regressão estão; O **RMSE** é uma medida de quão espalhados são esses resíduos. Em outras palavras, ele informa o quão concentrados os dados estão em torno da linha de melhor ajuste.



###  Por que MAE é melhor que RMSE?

Em sínteze,o **RMSE** tem o benefício de penalizar mais erros grandes, então pode ser mais apropriado em alguns casos, por exemplo, se errar por 10 é mais que duas vezes pior que errar por 5. Mas se errar por 10 é apenas duas vezes pior que sendo desligado por 5, então MAE é mais apropriado.

###  Qual é o alcance do RMSE?

Como os erros são elevados ao quadrado antes da média, o RMSE dá um peso relativamente alto para erros grandes. Isso significa que o RMSE é mais útil quando grandes erros são particularmente indesejáveis.Tal maneira que  tanto o **MAE** quanto o **RMSE** podem variar de 0 a  infinito . São pontuações orientadas negativamente: valores mais baixos são melhores.

###  O que é um bom RMSE e MAE?

Portanto, para um modelo ideal, **RMSE/MAE=0 e pontuação R2 = 1** , e todos os pontos residuais estão no eixo X.

Uma **Learning Curve** mostra a relação da pontuação de treinamento versus a pontuação de teste em uma validada cruzada para um estimador com um número de variável de amostras de treinamento. Essa visualização é normalmente usada para mostrar duas coisas:


- **Quanto o estimador se beneficia de mais dados (por exemplo, temos “dados suficientes” ou o estimador ficará melhor se usado de forma online).**


- **Se o estimador é mais sensível ao erro devido à variância versus erro devido ao viés.**

### 12.1. Análise de variância ou Analysis of variance  - ANOVA 

Por meio desenvolvimento da  análise da variabilidade dos resíduos com múltiplas variável(is) independentes possui algumas peculiaridades e requer atenção na interpretação. Em especial, na compreensão dos efeitos parciais e combinados das variáveis independentes sobre a dependente, ou seja, **é preciso verificar em que medida é possível inferir sobre a relação linear entre as variáveis na população a partir do que observamos na amostra.**



Nesse sentido, Análise  de  Variância - ANOVA **é um teste estatístico (parmétrico que  possui estimativas de parâmetros) utilizado para  verificar  se  as "MÉDIAS" de dois ou mais grupos são significativamente diferentes entre si**. Normalmente a ANOVA será aplicada para comparar mais de 2 grupos, assim como ANOVA é semelhante ao **teste t** também é utilizado nesse tipo de procedimento (com no máximo de dois níveis ou grupos), porém a ANOVA é indicada para análises com um “n” amostral superior a 30

**Ideias iniciais:**

<font color="#f000"> Amostras independentes – isso significa que uma observação não pode ser influenciada pela anterior ou pela próxima. Esse pressuposto garante que os dados sejam coletados aleatoriamente dentro do espaço amostral.</font>

<font color="#f000"> Homogeneidade das variâncias entre os grupos – as variância dentro de cada grupo é igual (ou pelo menos aproximadamente) àquela dentro de todos os grupos. Desta forma, cada tratamento contribui de forma igual para a soma dos quadrados.</font>

Além disso, os resíduos seguem uma distribuição normal assume-se que a média geral dos resíduos é igual a zero, ou seja, distribuem-se normalmente. Esse pressuposto não é muito restritivo dependendo do tamanho amostral.

De certo, o resultado da ANOVA será interpretado com base no valor-p.Valor-p é a probabilidade de que a estatística do teste assuma um valor extremo em relação ao valor observado quando H0 é verdadeira.Regra geral de interpretação(considerando o valor-p de 0.05):

- **Valor-p baixo(menor que 0.05):**

> Forte evidência empírica contra H0(rejeitamos a H0).Isso indica probabilidade inferior a 5% de que a H0 está correta.
       
- **Valor-p alto(maior que 0.05):**

> Pouca ou nenhuma evidência empírica contra H0(falhamos em rejeitar a H0).
     
- **Valor-p exatamente igual a 0.05**
> O pequisador decide se rejeita ou não a H0 ou se refaz o teste.


No entanto, ANOVA derivam-se duas importantes estatísticas da qualidade do ajuste da regressão linear: o coeficiente de determinação **R² e a estatística F**, seram mencionados nas secões a seguir.

- O “jogo” é o resultado do modelo.
- Os “jogadores” são os recursos (variáveis) incluídos no modelo.

É importante ressaltar que o que chamamos de “jogo” diz respeito a uma única  observação. Um jogo = uma observação. De fato, SHAP é sobre interpretabilidade local de um modelo preditivo.

#### Como funcionam os valores de Shapley?

O valor de Shapley é a média de todas as contribuições marginais para todas as coalizões possíveis. **O tempo de computação aumenta exponencialmente com o número de recursos. Uma solução para manter o tempo de computação gerenciável é computar contribuições para apenas algumas amostras das possíveis coalizões.**

---
#### Por que usamos o método de valor Shapley?

O valor de Shapley se aplica principalmente em situações em que as contribuições de cada ator **são** desiguais, mas cada jogador trabalha em cooperação uns com os outros para obter o ganho ou recompensa. O **valor de Shapley** garante que cada ator ganhe tanto ou mais quanto **teria** de agir de forma independente.

---
#### O que é regressão de Shapley?

**A regressão Shapley Value** é uma técnica para descobrir a importância relativa das variáveis preditoras na regressão linear . Sua principal aplicação é resolver uma fraqueza da **regressão linear**, que é que ela não é confiável quando as variáveis previstas são moderadamente a altamente **correlacionadas.**

### 13.1.  Ideias iniciais 

segundo (PRABHAKARAN, 2020) afirma que o **intervalo de confiança** é uma medida para quantificar a incerteza em uma estatística estimada (como a média) quando o verdadeiro parâmetro populacional é desconhecido.

A principal ideia acerca dos intervalos de confiança é adicionar informação  à estimativa encontrada por meio do estimador.

- Conhecida a distribuição do estimador propomos um intervalo de variação  para estimativa com uma precisão probalilística.
    
O procedimento é muito similar ao que empregamos para obter o tamaho de uma amostra, saiba mais neste artigo <a href="https://www.linkedin.com/pulse/tipos-de-amostragens-e-tamanho-da-amostra-usando-python-aline-marques/">Tipos de amostragens e Tamanho da amostra</a> para garantir uma margem de erro à estimativa com uma confiançã pré-determinada.

> O intervalo de confiança é uma técnica simples de inferência que permite estabelecer um intervalo com
probabilidade γ de conter um parâmetro populacional.Por exemplo, a especificação de um intervalo com confiança de 95% para 	${\infty}$**Y** significaria que este, em repetidas amostras de tamanho n, conteria o real rendimento médio da população em 95% das situações.

> Por exemplo, vamos supor que o tamanho da amostra é dado e que desejamos encontra um intervalo que aumente as chances de capturamos a verdadeira média de uma poulação.Assim como, **na prática nunca sabemos a média da população, temos exatamente informações das amostras, ou seja, nós esperamos dado um intervalo de confiança (o valor em %) das médias das amostras de uma população, irão produzir um intervalo de confiança que inclua a média da população**.


### 13.2. Diferença entre o parâmetro da população vs estatística da amostra



Segundo (PRABHAKARAN, 2020) explica que um parâmetro populacional (como média, desvio padrão, etc) é aquele que é calculado ou conhecido de toda a população. Considerando que, uma estatística de amostra é calculada a partir de uma amostra menor de toda a população. O parâmetro de população é difícil de obter/computar. Caso contrário, não estaríamos nos preocupando tanto com estatísticas. Além disso, **é aqui que os intervalos de confiança entram em cena**. Como muitas vezes não é prático calcular o parâmetro populacional, calculamos a estatística a partir de uma amostra menor e então estimamos um intervalo de confiança entre o qual o parâmetro populacional verdadeiro pode variar.



Percebemos que o intervalo de confiança refere-se a uma estimativa para um parâmetro populacional, uma constante,em poucas palavras na variabilidade da média amostral,  já por outro lado intervalo de previsão refere-se a uma estimativa para variáveis aleatórias, ou seja, na variabilidade dos valores individuais.

Portanto, calcularemos manualmente o intervalo de confiança para o RMSE

```python
confidence = 0.95 # define o coeficiente
squared_errors = (final_predictions - y_test) ** 2  # obtendo os erros ao quadráticos
m = len(squared_errors) # tamanho da amostra em dados de test
mean = squared_errors.mean() # calcula a média de erro

"""calcula o valor crítico t Table, ou seja, é a função de ponto percentual (inverso de 'cdf',  que é probabilidade acumulada) 

m-1 : é t-Student com n-1 graus de liberdade

"""
tscore = stats.t.ppf((1 + confidence) / 2, df=m - 1)  # teremos aqui 10 graus de liberdade na table t para varia

tmargin = tscore * squared_errors.std(ddof=1) / np.sqrt(m) # calcular a margem de erro do limite inferior e superior, ou seja, aqui que calcular o intervalo de confiança

# mostra limiter inferior e superior
ic = np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)

#view
ic
```

O motivo de usa a função **abs**,é porque ela facilita no cálcula da raiz, ou seja, como cálculo gerou número negativo no limiter inferior.Dessa forma, não conseguimos determinar a raiz de um número negativo, não há raízes reais, tal modo que, violamos umas das propriedade matemáticas em relação cálculo de raizes.

### 14. Deploy do modelo

vamos criar uma Pipeline com dois estágios:

- **drop_columns**: Remove as colunas indesejadas do conjunto de dados de entrada.
- **classification**: Alimenta um modelo de classificação com os dados obtidos no estágio **drop_columns**, podendo ser tanto para treinamento quanto para obter uma predição.
Para criar um modelo capaz de fazer transformações nos dados de entrada, vamos criar uma `Pipeline` do `scikit-learn` e aplicar nossas transformações dentro dos estágios dela.

Abaixo, definimos um transformador exemplo, que irá remover as colunas passadas como parâmetro em sua inicialização:

```python
# Um transformador para remove colunas indesejadas
class DropColumns(BaseEstimator, TransformerMixin):
    def __init__(self, columns):
        self.columns = columns

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        # Primero realizamos a cópia do Dataframe  'X' de entrada
        data = X.copy()
        
        # Retornamos um novo dataframe sem as colunas indesejadas
        return data.drop(labels=self.columns, axis=1)
```

Tanto o método `fit` quanto o `transform` devem obrigatoriamente ser definidos, mesmo se não forem fazer nada de diferente, como no caso do `fit` acima.

Da mesma forma, você pode criar outros transformadores, para outros propósitos, sempre herdando das classes `BaseEstimator` e `TransformerMixin`. Você pode utilizar um transformador para, por exemplo, criar novas colunas, editar tipos de dados de colunas existentes, entre outros.

Agora, vamos criar uma Pipeline para utilização do nosso modelo, aceitando todas as colunas esperadas pelo desafio e removendo as que não queremos usar.

```python
challenge_columns = df_copy.columns.tolist()

#features = X.columns.tolist()
features= ['Battery', 'Price', 'Torque', 'Permissible_weight']
target = ['Energy_consumption']

# Remove todas as colunas indesejadas
unwanted_columns = list((set(challenge_columns) - set(target)) - set(features))
unwanted_columns
```

Automatizando todo esse processo todo em único pipeline com a preparação completa dos dados e a previsão final.

#### Criação de um pipeline completo com preparação e previsão

```python
# Criando uma instância do transformador, passando como parâmetro as colunas que não queremos
drop_columns = DropColumns(unwanted_columns)

#Criando um pipeline completo com preparação e previsão
full_pipeline_with_predictor = Pipeline([
                         ('drop_cols',drop_columns),
                         ('scaler',MinMaxScaler()),
                         ('reg_XGB',reg_XGB)
])
full_pipeline_with_predictor
```
